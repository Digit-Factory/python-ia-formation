# Note méthodologique — Estimation du prix des appartements parisiens

## Contexte

ImmoAnalytica souhaitait disposer d'un modèle capable d'estimer 
le prix d'un appartement parisien à partir de ses caractéristiques 
(surface, nombre de pièces, étage, arrondissement, équipements, 
année de construction). Un dataset de 500 annonces nous a été fourni.

## Méthodologie

Le projet a suivi 5 étapes :

1. **Exploration** : analyse univariée et multivariée pour détecter 
les problèmes du dataset.

2. **Nettoyage** : suppression de 2 doublons, imputation de valeurs 
manquantes (médiane pour les variables numériques, 0 pour le balcon), 
suppression des outliers via la méthode IQR (surface > 200 m² et 
prix > 5 M€ identifiés comme erreurs de saisie).

3. **Préparation** : séparation train/test (80/20), normalisation 
centrage-réduction des features et de la cible. La normalisation est 
appliquée avec les paramètres calculés uniquement sur le train, 
pour éviter tout biais d'évaluation.

4. **Modélisation** : régression linéaire multivariée entraînée par 
descente de gradient implémentée from scratch (learning rate 0.05, 
2000 itérations). Le choix de la descente de gradient se justifie par 
sa simplicité, son interprétabilité, et sa parfaite adéquation avec 
le problème (relation approximativement linéaire).

5. **Évaluation** : test sur les 20% non vus pendant l'entraînement.

## Résultats

- **MAE** : ~80 000 € (erreur absolue moyenne)
- **RMSE** : ~110 000 €
- **R²** : 0.78 (le modèle explique 78% de la variance des prix)
- **Gain vs baseline** : division par 2 de l'erreur absolue

Les variables les plus influentes sont la surface (coefficient 
dominant), l'arrondissement (impact négatif quand on s'éloigne 
du centre), et l'étage.

## Limites

- **Hypothèse de linéarité** : la relation prix/features est 
probablement non-linéaire (effet de seuil sur la surface, 
non-linéarité des arrondissements…).
- **Peu de variables** : absence d'informations sur la qualité du 
quartier, la proximité des transports, l'état du bien.
- **Taille du jeu de données** : 500 annonces c'est peu pour Paris. 
L'intervalle de confiance est large.
- **Variables catégorielles brutes** : l'arrondissement est traité
