# Le vecteur : une flèche dans l'espace

## Intuition visuelle

Un **vecteur**, ce n'est rien d'autre qu'**une flèche**. Elle a :

- un **point de départ** (souvent l'origine (0,0))
- un **point d'arrivée**
- donc une **direction** et une **longueur**

En pratique, pour désigner un vecteur, on donne juste les coordonnées de son point d'arrivée (si on part de l'origine).

**Exemple** : le vecteur `v = (3, 2)` va de (0,0) à (3,2).

````python
#| fig-cap: "Le vecteur v = (3, 2) représenté comme une flèche"
fig, ax = plt.subplots()
ax.quiver(0, 0, 3, 2, angles='xy', scale_units='xy', scale=1, 
          color='red', width=0.01)
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 4)
ax.axhline(y=0, color='gray', linewidth=0.5)
ax.axvline(x=0, color='gray', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')
ax.set_title("Le vecteur v = (3, 2)")
ax.annotate('v = (3, 2)', xy=(3, 2), xytext=(3.2, 2.1), 
            fontsize=14, color='red')
plt.show()
````

## Deux manières de voir un vecteur

- Taille 100 : `(x1, x2, ..., x100)` — impossible à dessiner, mais parfaitement manipulable

# Opérations de base sur les vecteurs

## Addition de vecteurs

Quand on additionne deux vecteurs, on additionne **composante par composante** :

$$\vec{u} + \vec{v} = (u_1 + v_1, u_2 + v_2)$$

**Géométriquement** : c'est la règle du parallélogramme. Tu mets le début du 2ème vecteur à la fin du 1er.

````python
# Addition en Python pur
u = [2, 1]
v = [1, 3]

# Méthode naïve : boucle
somme = []
for i in range(len(u)):
    somme.append(u[i] + v[i])
print(f"u + v = {somme}")

# Méthode plus pythonique
somme_bis = [u[i] + v[i] for i in range(len(u))]
print(f"u + v = {somme_bis}")
````

````python
# Avec NumPy, c'est BEAUCOUP plus simple
u = np.array([2, 1])
v = np.array([1, 3])
print(f"u + v = {u + v}")
````

> **⚠️ Attention**
>
## ⚠️ Piège classique
En Python pur, `[2, 1] + [1, 3]` donne `[2, 1, 1, 3]` (concaténation !), **pas** l'addition. C'est une des raisons pour lesquelles NumPy existe.

````python
a = [2, 1]
b = [1, 3]
print(f"Listes Python : {a} + {b} = {a + b}  (concaténation !)")

a_np = np.array([2, 1])
b_np = np.array([1, 3])
print(f"Tableaux NumPy : {a_np} + {b_np} = {a_np + b_np}  (vraie addition)")
````

````python
#| fig-cap: "Addition de vecteurs : règle du parallélogramme"
fig, ax = plt.subplots()

u = np.array([2, 1])
v = np.array([1, 3])
somme = u + v

ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, 
          color='red', label=f'u = {u}')
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, 
          color='blue', label=f'v = {v}')
ax.quiver(0, 0, somme[0], somme[1], angles='xy', scale_units='xy', scale=1, 
          color='green', label=f'u+v = {somme}')
ax.quiver(u[0], u[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, 
          color='blue', alpha=0.3)

ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.axhline(y=0, color='gray', linewidth=0.5)
ax.axvline(x=0, color='gray', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')
ax.legend(loc='upper left')
ax.set_title("Addition de vecteurs : règle du parallélogramme")
plt.show()
````

## Multiplication par un scalaire

Multiplier un vecteur par un nombre, c'est **étirer** ou **compresser** la flèche :

$$k \cdot \vec{v} = (k \cdot v_1, k \cdot v_2)$$

- Si $k > 1$ : la flèche s'allonge
- Si $0 < k < 1$ : la flèche raccourcit
- Si $k < 0$ : la flèche change de sens

````python
v = np.array([2, 1])
print(f"v       = {v}")
print(f"2 * v   = {2 * v}   (doublé)")
print(f"0.5 * v = {0.5 * v} (moitié)")
print(f"-1 * v  = {-1 * v}  (sens inverse)")
````

# Le produit scalaire — LA notion clé

Accroche-toi, c'est LA notion à ne surtout pas louper. Elle est partout en IA.

## Définition

Le produit scalaire de deux vecteurs `u` et `v` (noté `u · v`) est :

$$\vec{u} \cdot \vec{v} = u_1 \times v_1 + u_2 \times v_2 + \dots + u_n \times v_n$$

**On multiplie composante par composante, puis on additionne tout.**

> **⚠️ Attention**
>
Le résultat est **UN seul nombre** (un scalaire), pas un vecteur !

````python
# Produit scalaire à la main
u = [1, 2, 3]
v = [4, 5, 6]

resultat = 0
for i in range(len(u)):
    resultat += u[i] * v[i]
print(f"u · v = {u[0]}×{v[0]} + {u[1]}×{v[1]} + {u[2]}×{v[2]} = {resultat}")
````

````python
# Avec NumPy : trois syntaxes équivalentes
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

print(f"np.dot(u, v) = {np.dot(u, v)}")
print(f"u @ v        = {u @ v}")
print(f"u.dot(v)     = {u.dot(v)}")
````

## Intuition : « à quel point ils pointent dans la même direction ? »

Le produit scalaire répond à la question : **« les deux vecteurs vont-ils dans la même direction ? »**

- Produit scalaire **grand et positif** → même direction 🔗
- Produit scalaire **proche de 0** → perpendiculaires ⊥
- Produit scalaire **négatif** → directions opposées ↔️

C'est ce qui permet, en IA, de mesurer la **similarité** entre deux vecteurs.

````python
#| fig-cap: "Trois cas d'interprétation du produit scalaire"
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

cas = [
    ("Même direction (ps > 0)", np.array([3, 1]), np.array([2, 1])),
    ("Perpendiculaires (ps ≈ 0)", np.array([3, 0]), np.array([0, 3])),
    ("Opposés (ps < 0)", np.array([3, 1]), np.array([-2, -1])),
]

for ax, (titre, u, v) in zip(axes, cas):
    ax.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, 
              color='red', label=f'u = {u}')
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, 
              color='blue', label=f'v = {v}')
    ps = np.dot(u, v)
    ax.set_title(f"{titre}\nu · v = {ps}")
    ax.set_xlim(-4, 4)
    ax.set_ylim(-3, 3)
    ax.axhline(y=0, color='gray', linewidth=0.5)
    ax.axvline(x=0, color='gray', linewidth=0.5)
    ax.grid(True, alpha=0.3)
    ax.set_aspect('equal')
    ax.legend(loc='upper left', fontsize=9)

plt.tight_layout()
plt.show()
````

# Norme et distance

## La norme : la longueur du vecteur

La **norme** d'un vecteur (notée $||v||$), c'est **sa longueur**. Formule (Pythagore généralisé) :

$$||\vec{v}|| = \sqrt{v_1^2 + v_2^2 + \dots + v_n^2}$$

On appelle ça aussi la **norme euclidienne** ou **norme L2**.

````python
v = np.array([3, 4])

norme_manuelle = np.sqrt(v[0]**2 + v[1]**2)
print(f"Norme de {v} à la main : {norme_manuelle}")

norme_numpy = np.linalg.norm(v)
print(f"Norme de {v} avec NumPy : {norme_numpy}")
````

## Distance entre deux vecteurs

La **distance euclidienne** entre deux points `a` et `b`, c'est la norme du vecteur `a - b` :

$$d(\vec{a}, \vec{b}) = ||\vec{a} - \vec{b}||$$

````python
a = np.array([1, 2])
b = np.array([4, 6])

distance = np.linalg.norm(a - b)
print(f"Distance euclidienne : {distance}")
````

## Autres distances utiles

| Distance | Formule | Intuition |
|---|---|---|
| **Manhattan (L1)** | $\sum_i \|a_i - b_i\|$ | « En taxi dans New York » |
| **Euclidienne (L2)** | $\sqrt{\sum_i (a_i - b_i)^2}$ | « À vol d'oiseau » |
| **Chebyshev (L∞)** | $\max_i \|a_i - b_i\|$ | Plus grande différence sur une dimension |

````python
a = np.array([1, 2])
b = np.array([4, 6])

d_manhattan = np.sum(np.abs(a - b))
d_euclid = np.linalg.norm(a - b)
d_cheby = np.max(np.abs(a - b))

print(f"Distance de Manhattan : {d_manhattan}")
print(f"Distance euclidienne  : {d_euclid:.2f}")
print(f"Distance de Chebyshev : {d_cheby}")
````

> **ℹ️ Note**
>
## 🧠 À retenir
Le choix de la distance influe sur le comportement des algorithmes (k-NN, k-means...). L'euclidienne est la plus fréquente, mais pas toujours la meilleure.


# Les matrices

## Définition

Une **matrice**, c'est un **tableau rectangulaire de nombres**. On parle de matrice de taille `m × n` (m lignes, n colonnes).

$$M = \begin{pmatrix} 1 & 2 \\ 3 & 4 \\ 5 & 6 \end{pmatrix}$$

````python
M = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

print("Matrice M :")
print(M)
print(f"\nTaille : {M.shape}  (lignes, colonnes)")
print(f"Accéder à M[1, 0] (2e ligne, 1re colonne) : {M[1, 0]}")
````

## Lien avec les données réelles

Un jeu de données **est** une matrice :

- Chaque **ligne** = un individu
- Chaque **colonne** = une variable

````python
donnees = np.array([
    [35, 42000, 2, 1],   # Client 1
    [52, 65000, 0, 1],   # Client 2
    [28, 31000, 1, 0],   # Client 3
])

print(f"Jeu de données {donnees.shape} :")
print(donnees)
print(f"\nClient 2 : {donnees[1]}")
print(f"Colonne salaire : {donnees[:, 1]}")
````

## Le produit matriciel — LE calcul central du ML

> **⚠️ Attention**
>
## ⚠️ Attention
Le produit matriciel **n'est pas** élément par élément. C'est un calcul plus subtil.


### Règle de base

Pour multiplier `A × B` :

- Le nombre de **colonnes de A** doit être égal au nombre de **lignes de B**
- Si A est `m × n` et B est `n × p`, alors `A × B` est `m × p`

### Intuition

Chaque élément de la matrice résultat est un **produit scalaire** :

> `(A×B)[i, j]` = produit scalaire entre la **ligne i de A** et la **colonne j de B**

````python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("A @ B =")
print(A @ B)
print()
print("Vérification du premier élément :")
print(f"(A×B)[0,0] = ligne 0 de A · colonne 0 de B")
print(f"         = [1,2] · [5,7] = 1×5 + 2×7 = {1*5 + 2*7}")
````

````python
# Attention : A @ B ≠ B @ A (non commutatif !)
print("A @ B =")
print(A @ B)
print("\nB @ A =")
print(B @ A)
````

## Pourquoi c'est partout en ML ?

Un **neurone** calcule : $y = w_1 x_1 + w_2 x_2 + \dots + w_n x_n + b$

Avec les notations vectorielles : $y = \vec{w} \cdot \vec{x} + b$

Une **couche de neurones** :

$$\vec{y} = W \vec{x} + \vec{b}$$

où `W` est une matrice de poids. **C'est un produit matriciel.**

> **💡 Astuce**
>
Tout modèle de deep learning, ce sont juste des produits matriciels enchaînés (avec des fonctions d'activation entre les deux). Voilà, c'est aussi simple et aussi profond que ça.

# Transposée, inverse, déterminant

On ne rentre pas dans les calculs à la main — juste à quoi ça sert et comment faire avec NumPy.

## La transposée (Aᵀ)

Transposer, c'est **échanger lignes et colonnes**.

````python
M = np.array([[1, 2, 3], [4, 5, 6]])
print(f"M (taille {M.shape}) :")
print(M)
print(f"\nM.T (taille {M.T.shape}) :")
print(M.T)
````

**À quoi ça sert ?** Rendre compatibles deux matrices dans un produit, et apparaît dans toutes les formules de ML.

## L'inverse (A⁻¹)

L'inverse d'une matrice carrée `A`, c'est la matrice `A⁻¹` telle que :

$$A \times A^{-1} = I$$

où `I` est la **matrice identité** (l'équivalent de « 1 » pour les matrices).

> **⚠️ Attention**
>
Toutes les matrices n'ont **pas** d'inverse.

````python
A = np.array([[4, 7], [2, 6]])

A_inv = np.linalg.inv(A)
print("A⁻¹ :")
print(A_inv)

print("\nA @ A⁻¹ doit donner la matrice identité :")
print(np.round(A @ A_inv, 10))
````

**À quoi ça sert ?** À résoudre des systèmes d'équations linéaires — la base de la régression linéaire.

## Le déterminant

Un nombre unique calculé à partir d'une matrice carrée.

- `det(A) = 0` → la matrice n'a pas d'inverse
- `det(A) ≠ 0` → la matrice a un inverse

````python
A = np.array([[4, 7], [2, 6]])
B = np.array([[2, 4], [1, 2]])  # lignes proportionnelles

print(f"det(A) = {np.linalg.det(A):.2f}")
print(f"det(B) = {np.linalg.det(B):.2f}  (proche de 0 → pas d'inverse)")
````
