
Cette notion te donne les bases pour **ne plus jamais faire ce genre d'erreur**, et pour choisir consciemment quel traitement appliquer à chaque variable.

# La grande carte des types de variables

## 🗺️ Vue d'ensemble

Il existe **4 grandes catégories** de variables, avec des sous-types :

```
                    Variables
                        │
      ┌─────────────────┼─────────────────┐
      │                 │                 │
  Quantitatives    Qualitatives       Textuelles
  (numériques)    (catégorielles)   (texte libre)
      │                 │
  ┌───┴───┐        ┌────┴────┐
  │       │        │         │
Continues Discrètes Nominales Ordinales
                                │
                           Temporelles
                           (dates, heures)
```

## 🔢 1. Variables quantitatives

**Ce qu'elles mesurent** : une **quantité numérique** sur laquelle les opérations mathématiques (somme, moyenne, différence) ont un **sens**.

Deux sous-types :

### 1a. Quantitatives continues

Peuvent prendre **n'importe quelle valeur** dans un intervalle.

**Exemples** :
- Taille (`1.67 m`)
- Poids (`68.3 kg`)
- Température (`24.5°C`)
- Prix (`49.99 €`)
- Chiffre d'affaires

**Dtype Pandas** : `float64` (ou `float32`)

### 1b. Quantitatives discrètes

Ne peuvent prendre que des valeurs **entières** (et chaque valeur a un sens de comptage).

**Exemples** :
- Nombre d'enfants
- Nombre de pièces d'un appartement
- Nombre de commandes par client
- Âge (en années entières)

**Dtype Pandas** : `int64` (ou `int32`)

> **⚠️ Attention**
>
## ⚠️ Piège
Tu as le droit de calculer la moyenne d'une variable discrète (`moyenne = 2.3 enfants`), **même si la moyenne n'est pas une valeur "atteignable"**. C'est une statistique pertinente.


## 🏷️ 2. Variables qualitatives (catégorielles)

**Ce qu'elles mesurent** : une **catégorie**, pas une quantité. Les opérations mathématiques n'ont **aucun sens**.

Deux sous-types :

### 2a. Qualitatives nominales

Catégories **sans ordre naturel**.

**Exemples** :
- Genre (Homme, Femme, Autre)
- Pays de résidence
- Couleur préférée (Rouge, Bleu, Vert)
- Catégorie de produit (Électronique, Vêtement, Alimentaire)

### 2b. Qualitatives ordinales

Catégories **avec un ordre naturel**.

**Exemples** :
- Niveau de satisfaction (Mauvais < Moyen < Bon < Excellent)
- Taille de vêtement (XS < S < M < L < XL)
- Niveau d'études (Bac < Licence < Master < Doctorat)
- Classement (1er, 2e, 3e…)

**Dtype Pandas** : `object` (string) ou `category` (plus efficace en mémoire)

> **💡 Astuce**
>
## 🧠 La différence entre nominal et ordinal est cruciale
Pour les **nominales**, tu utilises **One-Hot Encoding** (une colonne par modalité).  
Pour les **ordinales**, tu peux utiliser **Label Encoding** (un entier croissant) parce que l'ordre a du sens.

On verra ça en détail dans la Notion 3.4.


### Le piège des "codes numériques"

Certaines variables **semblent** numériques mais sont en réalité nominales :

- **Code postal** (`75001`, `13000`…)
- **Identifiant client** (`12345`, `67890`…)
- **Numéro de département** (`75`, `13`, `69`…)

Une moyenne de code postal n'a aucun sens, même si Pandas peut la calculer !

## 📝 3. Variables textuelles

**Ce qu'elles contiennent** : du **texte libre**, non catégorisé.

**Exemples** :
- Commentaire client
- Description d'un produit
- Article de blog
- Tweet

**Dtype Pandas** : `object`

La différence avec les variables qualitatives ? Les **textuelles** n'ont pas un petit nombre de modalités répétées. Chaque valeur est (quasi) unique.

On les traite avec des techniques de **NLP** (traitement du langage naturel) : tokenisation, embedding… On verra ça au Module 7.

## 📅 4. Variables temporelles

**Ce qu'elles contiennent** : des **dates**, des **heures**, des **durées**.

**Exemples** :
- Date de naissance
- Date d'achat
- Heure d'une transaction
- Durée d'une session

**Dtype Pandas** : `datetime64[ns]`

Elles ont des propriétés spéciales : on peut extraire l'année, le mois, le jour de la semaine…

## 📊 Tableau récapitulatif

| Catégorie | Sous-type | Dtype | Exemple | Moyenne ? | Ordre ? |
|---|---|---|---|:---:|:---:|
| Quantitative | Continue | `float64` | Taille | ✅ | ✅ |
| Quantitative | Discrète | `int64` | Nb enfants | ✅ | ✅ |
| Qualitative | Nominale | `object` / `category` | Pays | ❌ | ❌ |
| Qualitative | Ordinale | `object` / `category` | Satisfaction | ❌ | ✅ |
| Textuelle | — | `object` | Commentaire | ❌ | ❌ |
| Temporelle | — | `datetime64` | Date | ❌ | ✅ |

## ⚠️ Le danger des "codes numériques"

Regarde ce qui se passe si je fais un `describe()` par défaut :

````python
df.describe()
````

**Pandas a calculé la moyenne de `code_postal` et `id_client` !** Ces statistiques n'ont **aucun sens**. Ton job est de **reconnaître ce problème** et de corriger.

## 🔧 Corriger le type d'une variable

### Convertir en string / category

Pour signaler à Pandas qu'une variable numérique est en fait catégorielle :

````python
# Convertir en string (simple)
df["code_postal"] = df["code_postal"].astype(str)
df["id_client"] = df["id_client"].astype(str)

print(df.dtypes)
````

Maintenant, `describe()` ne calcule plus la moyenne de ces colonnes :

````python
df.describe()
````

Seulement les vraies variables numériques ! ✅

### Utiliser le type `category` pour les qualitatives

Le type `category` est **plus efficace en mémoire** et permet certaines opérations :

````python
df["ville"] = df["ville"].astype("category")
df["genre"] = df["genre"].astype("category")

print(df[["ville", "genre"]].dtypes)
````

````python
# Gain de mémoire (surtout sur gros datasets)
memoire_object = df["ville"].astype("object").memory_usage(deep=True)
memoire_category = df["ville"].memory_usage(deep=True)
print(f"Mémoire avec object   : {memoire_object} octets")
print(f"Mémoire avec category : {memoire_category} octets")
````

Sur un petit dataset la différence est minime, mais sur des millions de lignes **c'est énorme**.

### Encoder l'ordre d'une variable ordinale

Pour signaler à Pandas que `satisfaction` a un ordre, on utilise le type `CategoricalDtype` :

````python
from pandas.api.types import CategoricalDtype

# Définir l'ordre
ordre_satisfaction = CategoricalDtype(
    categories=["Mauvais", "Moyen", "Bon", "Excellent"],
    ordered=True
)
df["satisfaction"] = df["satisfaction"].astype(ordre_satisfaction)

print(f"Dtype : {df['satisfaction'].dtype}")
print(f"Ordre : {df['satisfaction'].cat.categories.tolist()}")
````

Maintenant, les comparaisons ont du sens :

````python
# Qui a eu au moins "Bon" ?
df[df["satisfaction"] >= "Bon"][["ville", "satisfaction"]]
````

C'est magique : une comparaison `>=` sur du texte fonctionne parce qu'on a défini l'ordre.

# Exploration par type de variable

## 🎯 Quelles analyses pour quel type ?

Selon le type, les analyses pertinentes changent :

| Type | Statistiques | Visualisations |
|---|---|---|
| **Quantitative continue** | Moyenne, médiane, écart-type, quantiles | Histogramme, boxplot, scatter |
| **Quantitative discrète** | Idem + comptage des valeurs | Bar chart (petites valeurs), histogramme |
| **Qualitative nominale** | Modalités, fréquences, mode | Bar chart, camembert (max 5 catégories) |
| **Qualitative ordinale** | Idem + médiane (si ordre numérique) | Bar chart ordonné |
| **Textuelle** | Longueur moyenne, nb mots, top tokens | Word cloud, histogramme longueur |
| **Temporelle** | Plage, fréquence, saisonnalité | Line plot, heatmap calendrier |

## 🔢 Explorer une variable quantitative

````python
# Exemple : poids
print(df["poids_kg"].describe())
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Histogramme
axes[0].hist(df["poids_kg"], bins=6, color="steelblue", edgecolor="black", alpha=0.7)
axes[0].axvline(df["poids_kg"].mean(), color="red", linestyle="--", label=f"Moyenne")
axes[0].axvline(df["poids_kg"].median(), color="green", linestyle="--", label=f"Médiane")
axes[0].set_title("Distribution du poids")
axes[0].set_xlabel("Poids (kg)")
axes[0].legend()

# Boxplot
axes[1].boxplot(df["poids_kg"], vert=True)
axes[1].set_title("Boxplot du poids")
axes[1].set_ylabel("Poids (kg)")

plt.tight_layout()
plt.show()
````

## 🏷️ Explorer une variable qualitative nominale

````python
# Comptage des modalités
print(df["ville"].value_counts())
````

````python
fig, ax = plt.subplots(figsize=(8, 4))
df["ville"].value_counts().plot(kind="bar", ax=ax, color="coral", edgecolor="black")
ax.set_title("Répartition des villes")
ax.set_ylabel("Nombre d'observations")
ax.tick_params(axis="x", rotation=0)
plt.tight_layout()
plt.show()
````

## 📊 Explorer une variable qualitative ordinale

Pour une ordinale, on conserve **l'ordre** dans le graphique :

````python
fig, ax = plt.subplots(figsize=(8, 4))
df["satisfaction"].value_counts().sort_index().plot(
    kind="bar", ax=ax, color="mediumseagreen", edgecolor="black"
)
ax.set_title("Niveau de satisfaction (ordonné)")
ax.set_ylabel("Nombre")
ax.tick_params(axis="x", rotation=0)
plt.tight_layout()
plt.show()
````

Remarque : **ne trie jamais par fréquence** une ordinale, car ça casse la lecture chronologique/logique.

# Pièges classiques d'identification

## 🚨 Piège 1 — Les variables binaires (0/1)

Une variable codée en 0/1 est **qualitative nominale**, pas quantitative !

**Exemples** : `a_enfants` (1 si oui), `est_client_premium`, `fume` (1/0).

La moyenne d'une variable binaire a une **signification particulière** : c'est la **proportion de 1**.

```python
df["est_client_premium"].mean()  # → 0.23 signifie 23% de clients premium
```

Donc ce n'est **pas la même chose** qu'une moyenne classique. On la traite comme une qualitative nominale, mais la "moyenne" s'interprète comme une proportion.

## 🚨 Piège 2 — L'âge

**Débat classique** : l'âge est-il continu ou discret ?

- **En années entières** (ex: 25 ans) → **discret**
- **En années avec décimales** (ex: 25.3 ans) → **continu**

En pratique, on le traite souvent comme **continu** parce que les stats/graphiques sont plus riches.

## 🚨 Piège 3 — Les notes et évaluations

Une note de 0 à 20 est-elle quantitative ou ordinale ?

**Réponse nuancée** :

- Si les écarts **ont du sens** (un étudiant à 16 est 2 points au-dessus d'un à 14) → **quantitative discrète**
- Si on considère juste que c'est un rang (le 16 est mieux que le 14, mais on ne peut pas dire "de combien") → **ordinale**

En pratique, pour les notes scolaires ou les évaluations numériques à large amplitude, on les traite comme **quantitatives**.  
Pour les notes à petite amplitude (1 à 5 étoiles), on les traite plutôt comme **ordinales**.

## 🚨 Piège 4 — Les identifiants

**Tout identifiant est qualitatif nominal**, jamais quantitatif, même s'il est numérique.

- `id_client` = 12345
- `id_produit` = 789
- `code_postal` = 75015
- `numero_siret` = 44306184100047

La règle : **si tu ne calculerais jamais la moyenne dans un tableau Excel, c'est une qualitative.**

# 🎓 Récapitulatif

## 📋 Les 4 grandes catégories

| Catégorie | Sous-types | Dtype Pandas |
|---|---|---|
| **Quantitative** | Continue, Discrète | `float64`, `int64` |
| **Qualitative** | Nominale, Ordinale | `object`, `category` |
| **Textuelle** | — | `object` |
| **Temporelle** | — | `datetime64[ns]` |

## 🧠 Les 5 questions à te poser pour chaque variable

1. **Est-ce une quantité mesurable ?** → Si oui : quantitative, sinon qualitative
2. **Peut-elle prendre des valeurs décimales ?** → Continue si oui, discrète sinon
3. **Y a-t-il un ordre naturel ?** → Ordinale si oui, nominale sinon
4. **Est-ce un identifiant ou un code ?** → **Toujours qualitative nominale**, même si c'est un nombre
5. **Calculer la moyenne aurait-il un sens ?** → Si non, **ce n'est pas une quantitative**

## 🔧 Les conversions à retenir

```python
df["col"].astype(str)                    # vers string
df["col"].astype("category")             # vers category (économe en mémoire)
pd.to_datetime(df["col"])                # vers datetime
df["col"].astype(
    pd.CategoricalDtype(categories=[...], ordered=True)
)                                        # vers ordinale ordonnée
```
