## 🎯 Pourquoi ce module ?

Tu as vu au Module 2 comment **utiliser les outils** (NumPy, Pandas, Matplotlib). Maintenant on va **structurer ta démarche** de data scientist.

En entreprise, **70 à 80% du temps d'un projet de ML est consacré à la préparation des données**. Ce n'est pas une exagération, c'est une réalité. Les data scientists juniors négligent cette partie et se précipitent sur les modèles. Les seniors savent qu'un modèle médiocre sur des données propres bat toujours un modèle génial sur des données sales.

Ce module t'apprend à faire cette préparation **avec méthode et rigueur**.

## 📋 Ce que tu sauras faire à la fin

- Identifier **précisément** le type statistique de chaque variable
- Détecter **toutes les anomalies** d'un dataset (NaN, doublons, outliers, incohérences)
- Choisir la **bonne stratégie** pour chaque problème
- Préparer proprement des données pour un modèle de ML
- **Documenter** ton nettoyage pour que ton travail soit reproductible

## 💡 Approche pédagogique

Ce module **formalise** ce que tu as fait de manière intuitive au Module 2. On y introduit :

- Des **règles claires** pour identifier les types de variables
- Des **stratégies standardisées** pour chaque cas de nettoyage
- Le **vocabulaire** qu'utilisent les data scientists en entreprise

Ce vocabulaire commun est important : il te permettra de communiquer avec d'autres data scientists, de lire la littérature technique, et de mieux comprendre ce que font les bibliothèques comme scikit-learn.
