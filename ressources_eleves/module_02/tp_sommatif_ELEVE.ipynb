# Note méthodologique — Analyse du catalogue Netflix

## 1. Contexte

Cette analyse porte sur un dataset de ~8000 titres du catalogue Netflix 
(films et séries TV), avec leurs caractéristiques : type, pays, date 
d'ajout, durée, genres, classification d'âge. L'objectif : comprendre 
la structure et l'évolution du catalogue pour aider la direction 
éditoriale dans ses choix stratégiques.

## 2. Méthodologie

Le travail a suivi 4 étapes :

**Nettoyage** : suppression des doublons (15), conversion de la date 
d'ajout en type datetime (après nettoyage des espaces parasites), 
imputation des valeurs manquantes par "Unknown" pour director/cast/country, 
et suppression des rares lignes sans date_added. Les outliers de durée 
(films > 300 min, manifestement erronés) ont été passés en NaN.

**Feature engineering** : création des colonnes `duration_min` (films), 
`n_seasons` (séries), `year_added`, `month_added`, `year_gap` (délai 
entre sortie et ajout Netflix). Création de DataFrames "explosés" pour 
les colonnes multi-valeurs (genres, pays) afin de permettre les 
agrégations correctes.

**Analyses** : KPI globaux, évolution temporelle, top genres et pays, 
distribution des durées, classifications.

**Visualisation** : dashboard synthétique 2×3 regroupant les 6 insights 
principaux pour présentation à la direction.

## 3. Principaux résultats

- **Catalogue** : ~75% de films et 25% de séries
- **Explosion 2015-2020** : le catalogue a été constitué principalement 
  pendant cette période, avec un ralentissement récent
- **Dominance USA/Inde/UK** : le trio de tête représente une majorité 
  du catalogue ; certains pays se spécialisent (Japon → séries, 
  Inde → films)
- **Durée médiane des films** : ~100 minutes, avec 80% des films entre 
  80 et 130 minutes
- **Genres en croissance** : les dramas et documentaires dominent 
  les ajouts récents

## 4. Limites

- Le dataset ne contient pas les **notes** des utilisateurs ni les 
  **audiences**, ce qui limite l'analyse qualitative
- Les **pays et genres multi-valeurs** compliquent certaines analyses 
  (un titre "USA, UK" est-il plutôt américain ou britannique ?)
- Les **classifications d'âge** sont hétérogènes entre pays 
  (TV-MA vs R vs autres systèmes)
- Pas d'information sur le **coût de production** ou le **succès** 
  des titres

## 5. Recommandations

1. **Enrichir le dataset** avec des données de performance (vues, 
   durée de visionnage, notes utilisateurs)
2. **Explorer la spécialisation géographique** : certains pays 
   semblent sur-performer sur certains types de contenus
3. **Analyser le cycle de vie** : les titres ajoutés récemment 
   performent-ils mieux que les anciens ? (nécessite données de vue)
4. **Étudier le délai Netflix-sortie** : impact sur la performance ?
5. **Segmentation par classification d'âge** : quelles tranches sont 
   sous-représentées ?

Bien que descriptive, cette analyse fournit une base solide pour 
orienter les investissements éditoriaux futurs. Les étapes suivantes 
devraient inclure des données de performance pour passer du descriptif 
au prescriptif.
