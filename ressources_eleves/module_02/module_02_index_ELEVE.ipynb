## 🎯 Pourquoi ce module ?

Au Module 1, tu as appris **les maths** du machine learning et tu as construit ton premier modèle à la main. Maintenant, il est temps de **consolider les outils** que tu vas utiliser tous les jours pendant toute ta carrière.

Ce module te transforme en **utilisateur·rice fluide** de l'écosystème data Python :

- **NumPy** pour les calculs numériques vectorisés
- **Pandas** pour manipuler des tableaux nommés (DataFrame)
- **Matplotlib / Seaborn** pour des visualisations propres et percutantes

Ce n'est pas un module "théorique" — c'est un module **outil**. À la fin, tu dois savoir manipuler n'importe quel dataset sans hésitation.
