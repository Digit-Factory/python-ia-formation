# Series et DataFrame : les deux structures fondamentales

## 🔢 La Series : un tableau 1D avec un index

Une **Series** c'est comme un vecteur NumPy, **mais avec un index nommé**. C'est l'équivalent d'une colonne de tableur.

````python
# Créer une Series
ventes_par_mois = pd.Series(
    [120, 150, 180, 210, 190, 220, 300, 280, 240, 260, 350, 400],
    index=["Jan", "Fév", "Mar", "Avr", "Mai", "Juin", 
           "Juil", "Août", "Sep", "Oct", "Nov", "Déc"],
    name="ventes_mensuelles"
)
print(ventes_par_mois)
````

**Avantages sur un array NumPy** :
- Accès par nom : `ventes_par_mois['Juin']` au lieu de `[5]`
- Opérations conservent l'index
- Méthodes intégrées : `.mean()`, `.sort_values()`, `.plot()`...

````python
print(f"Moyenne : {ventes_par_mois.mean():.1f}")
print(f"Ventes en juin : {ventes_par_mois['Juin']}")
print(f"\nMois où on a dépassé 250 :")
print(ventes_par_mois[ventes_par_mois > 250])
````

## 📊 Le DataFrame : un tableau 2D

Un **DataFrame** c'est un tableur complet : **plusieurs colonnes**, chacune étant une Series. C'est la structure que tu utiliseras 95% du temps.

````python
# Créer un DataFrame depuis un dictionnaire
df = pd.DataFrame({
    "ville": ["Paris", "Lyon", "Marseille", "Bordeaux", "Lille"],
    "population": [2_148_000, 513_000, 862_000, 254_000, 232_000],
    "surface_km2": [105, 48, 240, 49, 35],
    "region": ["IDF", "ARA", "PACA", "NAQ", "HDF"]
})

print(df)
````

````python
# Calculer une nouvelle colonne (densité)
df["densite"] = df["population"] / df["surface_km2"]
df["densite"] = df["densite"].round(0).astype(int)
print(df)
````

## 🎯 Quand Series vs DataFrame ?

| Structure | Cas d'usage |
|---|---|
| **Series** | Une seule colonne, une série temporelle simple, résultat d'agrégation |
| **DataFrame** | Jeu de données tabulaire (CSV, table SQL, Excel...) |

`df[["colonne"]]` (double crochets) te renvoie un **DataFrame à 1 colonne**.

# Charger des fichiers : l'entrée de toute analyse

## 📂 Les lecteurs `read_*`

Pandas lit **quasiment tous les formats** avec la même interface :

```python
pd.read_csv("fichier.csv")       # CSV
pd.read_excel("fichier.xlsx")    # Excel
pd.read_json("fichier.json")     # JSON
pd.read_parquet("fichier.pq")    # Parquet (format optimisé)
pd.read_html("http://...")       # Tableaux HTML d'une page web (!)
pd.read_sql(query, connection)   # Base de données
```

Tous retournent un **DataFrame**.

## 🔧 Options essentielles de `read_csv`

```python
pd.read_csv(
    "fichier.csv",
    sep=",",              # séparateur (, ; tab...)
    decimal=".",          # séparateur décimal (. en anglais, , en français)
    encoding="utf-8",     # encodage (ou "latin-1", "cp1252" pour fichiers français)
    parse_dates=["date"], # convertit automatiquement en dates
    na_values=["N/A", "-"], # valeurs à traiter comme NaN
    nrows=1000,           # lire seulement les 1000 premières lignes
    skiprows=2,           # ignorer les 2 premières lignes (commentaires)
)
```

> **⚠️ Attention**
>
## ⚠️ Encodage : le piège numéro 1
Les fichiers CSV français sont souvent en `latin-1` ou `cp1252`, pas en UTF-8. Si tu vois des caractères bizarres (`é` → `é`), essaie `encoding="latin-1"`.

## 🧹 Trois stratégies de traitement

### Stratégie 1 : supprimer les lignes concernées

````python
df_sans_nan = df.dropna()
print(f"Avant : {len(df)} lignes")
print(f"Après dropna() : {len(df_sans_nan)} lignes")
````

> **⚠️ Attention**
>
`dropna()` par défaut supprime **toute ligne qui a au moins un NaN**. Radical, mais parfois tu perds trop de données.

````python
# Supprimer uniquement sur certaines colonnes
df_clean = df.dropna(subset=["region"])
print(f"Sans region manquante : {len(df_clean)} lignes")
````

### Stratégie 2 : remplacer par une valeur

````python
# Remplacer les NaN de region par "Inconnue"
df_imputed = df.copy()
df_imputed["region"] = df_imputed["region"].fillna("Inconnue")

# Vérifier
print(df_imputed["region"].value_counts())
````

### Stratégie 3 : imputer intelligemment

````python
# Imputer par la modalité la plus fréquente
mode_region = df["region"].mode()[0]
df["region"] = df["region"].fillna(mode_region)

mode_paiement = df["mode_paiement"].mode()[0]
df["mode_paiement"] = df["mode_paiement"].fillna(mode_paiement)

print(f"NaN restants : {df.isna().sum().sum()}")
````

> **💡 Astuce**
>
## 🧠 Quelle stratégie choisir ?

| Situation | Stratégie |
|---|---|
| Peu de NaN et données abondantes | `dropna()` |
| NaN sur colonne numérique | `fillna(moyenne)` ou `fillna(médiane)` |
| NaN sur colonne catégorielle | `fillna(mode)` ou `"Inconnue"` |
| Beaucoup de NaN sur plusieurs colonnes | Imputation avancée (KNN, modèle...) |

# GroupBy : le cœur de l'analyse

**`GroupBy`**, c'est ce qui transforme un simple tableau en vraies analyses business.

## 🎯 Le principe "Split-Apply-Combine"

1. **Split** : on coupe les données par groupes
2. **Apply** : on applique une fonction à chaque groupe (somme, moyenne...)
3. **Combine** : on recombine les résultats

## 📊 Exemple 1 : CA par catégorie

````python
# 1 ligne pour répondre : quel CA par catégorie de produit ?
ca_par_cat = df.groupby("categorie")["montant_total"].sum().sort_values(ascending=False)
print(ca_par_cat)
````

## 📊 Exemple 2 : plusieurs statistiques d'un coup

````python
# Pour chaque catégorie : somme, moyenne, count
stats_cat = df.groupby("categorie")["montant_total"].agg(["sum", "mean", "count"])
print(stats_cat.round(2))
````

## 📊 Exemple 3 : groupby sur plusieurs colonnes

````python
# CA par catégorie ET région
ca_cat_region = df.groupby(["categorie", "region"])["montant_total"].sum()
print(ca_cat_region.head(10))
````

## 📊 Exemple 4 : différentes fonctions par colonne

````python
# Agrégations différentes selon la colonne
result = df.groupby("categorie").agg(
    ca_total=("montant_total", "sum"),
    ca_moyen=("montant_total", "mean"),
    nb_ventes=("id_transaction", "count"),
    quantite_totale=("quantite", "sum")
).round(2)

print(result)
````

# Jointures : combiner plusieurs tables

## 🔗 `merge` : le JOIN de SQL

Imagine qu'on a deux tables :
- **Ventes** : les transactions (celle qu'on a)
- **Catalogue** : les infos détaillées des produits (avec les fournisseurs)

On veut savoir **quel fournisseur a généré le plus de CA**. Pour ça, il faut **joindre** les deux tables sur `id_produit`.

````python
# Charger le catalogue
catalogue = pd.read_csv("ressources_tp/catalogue_produits.csv")
print(catalogue.head())
````

````python
# Joindre les deux tables sur id_produit
df_enrichi = df.merge(
    catalogue[["id_produit", "fournisseur"]],
    on="id_produit",
    how="left"  # garde toutes les lignes de df, même sans correspondance
)

print(f"Avant merge : {df.shape}")
print(f"Après merge : {df_enrichi.shape}")
print(df_enrichi[["id_produit", "nom_produit", "fournisseur"]].head())
````

## 🎯 Les 4 types de `how`

````python
# Exemple simple pour visualiser les 4 types de jointures
t1 = pd.DataFrame({"id": [1, 2, 3], "val_a": ["A", "B", "C"]})
t2 = pd.DataFrame({"id": [2, 3, 4], "val_b": ["X", "Y", "Z"]})
print("Table 1 :")
print(t1)
print("\nTable 2 :")
print(t2)
````

````python
# inner : seulement les lignes présentes dans LES DEUX tables
print("inner :")
print(t1.merge(t2, on="id", how="inner"))
````

````python
# left : toutes les lignes de la table de gauche
print("\nleft :")
print(t1.merge(t2, on="id", how="left"))
````

````python
# right : toutes les lignes de la table de droite
print("\nright :")
print(t1.merge(t2, on="id", how="right"))
````

````python
# outer : toutes les lignes des deux tables
print("\nouter :")
print(t1.merge(t2, on="id", how="outer"))
````

## 📚 Maintenant avec la jointure enrichie

````python
# CA par fournisseur
ca_fournisseur = (
    df_enrichi.groupby("fournisseur")["montant_total"]
    .sum().sort_values(ascending=False)
)
print(ca_fournisseur.round(0))
````

En **2 commandes** (`merge` + `groupby`), on a répondu à une question business complexe.

## 🧱 `concat` : empiler des DataFrames

````python
# Simuler deux "mois" de ventes
jan = df[df["date"].dt.month == 1]
fev = df[df["date"].dt.month == 2]

# Empiler verticalement
combine = pd.concat([jan, fev], axis=0)
print(f"Janvier : {len(jan)}, Février : {len(fev)}, Combiné : {len(combine)}")
````

## 🔄 `pivot_table` : Excel-style

````python
# Tableau croisé : CA par catégorie × région
pivot = df.pivot_table(
    values="montant_total",
    index="categorie",
    columns="region",
    aggfunc="sum",
    fill_value=0
).round(0)

print(pivot)
````

C'est l'équivalent du **tableau croisé dynamique** d'Excel.

# Les dates : une force de Pandas

Pandas gère les dates avec une élégance que peu d'autres outils peuvent offrir.

## 📅 `.dt` : les accessors de date

````python
# Extraire des composantes de la date
df["annee"] = df["date"].dt.year
df["mois"] = df["date"].dt.month
df["jour_semaine"] = df["date"].dt.day_name()
df["trimestre"] = df["date"].dt.quarter

print(df[["date", "annee", "mois", "jour_semaine", "trimestre"]].head())
````

## 📈 Resample : agréger par période de temps

````python
# Définir la date comme index pour utiliser resample
df_ts = df.set_index("date").sort_index()

# CA par semaine
ca_hebdo = df_ts["montant_total"].resample("W").sum()
print(ca_hebdo.head(10).round(0))
````

**Fréquences utiles** : `D` (jour), `W` (semaine), `ME` (fin de mois), `QE` (trimestre), `YE` (année).

## 📊 Visualisation rapide

````python
#| fig-cap: "Évolution hebdomadaire du CA"
fig, ax = plt.subplots(figsize=(12, 4))
ca_hebdo.plot(ax=ax, marker="o", linewidth=1.5)
ax.set_title("Chiffre d'affaires hebdomadaire")
ax.set_ylabel("CA (€)")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

On voit clairement le pic de fin d'année (Black Friday + Noël).

| Sélection par position | `.iloc[rows, cols]` |
| Gérer NaN | `.isna()`, `.fillna(x)`, `.dropna()` |
| Supprimer doublons | `.drop_duplicates()` |
| Grouper | `.groupby(col).agg(...)` |
| Joindre | `df1.merge(df2, on='col', how='left')` |
| Empiler | `pd.concat([...])` |
| Tableau croisé | `.pivot_table(values=..., index=..., columns=...)` |
| Dates | `.dt.year`, `.dt.month`, `.resample('W')` |

## 🧠 Les réflexes à prendre

1. **Toujours inspecter avant d'analyser** : `.head()`, `.info()`, `.isna().sum()`
2. **Nettoyer avant d'agréger** : doublons, NaN, types incohérents
3. **GroupBy** dès que tu veux faire une stat "par quelque chose"
4. **Merge** dès qu'il y a plusieurs tables à croiser
5. **Dates** : utilise `parse_dates` dès le chargement pour pouvoir utiliser `.dt`
