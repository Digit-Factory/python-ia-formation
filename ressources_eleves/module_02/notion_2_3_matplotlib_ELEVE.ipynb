# Comprendre Matplotlib : Figure, Axes, Artist

## 🏗️ L'architecture en 3 niveaux

Matplotlib paraît compliqué parce qu'il mélange **deux interfaces** : `plt.xxx()` (rapide) et orientée objet (propre). On va utiliser la seconde, qui est **la norme professionnelle**.

L'architecture :

| Concept | Description | Analogie |
|---|---|---|
| **Figure** | La fenêtre globale | Une feuille de papier |
| **Axes** | Une zone de dessin | Un graphique sur la feuille |
| **Artist** | Les éléments (ligne, barre, texte, titre...) | Les traits de crayon |

Une Figure peut contenir plusieurs Axes (subplots).

## 🎨 Le template à retenir

**95% de tes graphiques suivront ce squelette** :

```python
#| fig-cap: "Le squelette d'un graphique Matplotlib"

# 1. Créer une Figure et un Axes
fig, ax = plt.subplots(figsize=(8, 4))

# 2. Tracer quelque chose
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), label="sin(x)", color="steelblue", linewidth=2)

# 3. Habiller le graphique
ax.set_title("Un graphique bien habillé")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)")
ax.legend()
ax.grid(True, alpha=0.3)

# 4. Afficher
plt.show()
```

Les 4 étapes : **créer** → **tracer** → **habiller** → **afficher**.

## 🔢 Plusieurs sous-graphiques : `subplots(n_rows, n_cols)`

```python
#| fig-cap: "Plusieurs graphiques dans une seule figure"

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

x = np.linspace(0, 10, 100)

# Graphique de gauche
axes[0].plot(x, np.sin(x), color="steelblue")
axes[0].set_title("sin(x)")
axes[0].grid(True, alpha=0.3)

# Graphique de droite
axes[1].plot(x, np.cos(x), color="coral")
axes[1].set_title("cos(x)")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()  # évite les chevauchements
plt.show()
```

> **💡 Astuce**
>
## 🧠 À retenir
- `fig, ax = plt.subplots()` pour **1 graphique**
- `fig, axes = plt.subplots(2, 3)` pour **6 graphiques** (grille 2×3, `axes` est un array 2D)
- Toujours finir par `plt.tight_layout()` et `plt.show()`

# Les 6 graphiques essentiels en data science

Pour chaque type, on va voir **quand l'utiliser** et comment le produire proprement.

## 1️⃣ Scatter plot — relations entre deux variables

**Quand l'utiliser ?** Pour explorer la relation entre deux variables numériques.

```python
#| fig-cap: "Relation entre quantité et montant total"

fig, ax = plt.subplots(figsize=(10, 5))

ax.scatter(df["quantite"], df["montant_total"], 
           alpha=0.4, s=30, color="steelblue", edgecolor="k", linewidth=0.3)

ax.set_xlabel("Quantité")
ax.set_ylabel("Montant total (€)")
ax.set_title("Plus on achète, plus on paie... évident ?")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**Éléments à noter** :
- `alpha=0.4` : transparence pour voir la densité (sans ça, on ne voit qu'un blob noir)
- `s=30` : taille des points
- `edgecolor="k"` : bord noir, esthétique

## 2️⃣ Line plot — évolution dans le temps

**Quand l'utiliser ?** Pour montrer une évolution temporelle, des tendances, des séries.

```python
#| fig-cap: "Évolution du CA mensuel"

# Préparer les données
ca_mensuel = df.groupby(df["date"].dt.to_period("M"))["montant_total"].sum()

fig, ax = plt.subplots(figsize=(12, 4))

ax.plot(ca_mensuel.index.astype(str), ca_mensuel.values, 
        marker="o", linewidth=2, color="steelblue", markersize=8)

ax.set_xlabel("Mois")
ax.set_ylabel("CA (€)")
ax.set_title("Évolution du chiffre d'affaires mensuel en 2024")
ax.grid(True, alpha=0.3)

# Annoter le pic
mois_pic = ca_mensuel.idxmax()
ca_pic = ca_mensuel.max()
ax.annotate(f"Pic : {ca_pic:,.0f} €",
            xy=(str(mois_pic), ca_pic),
            xytext=(0.3, 0.9), textcoords="axes fraction",
            arrowprops=dict(arrowstyle="->", color="red"),
            fontsize=11, color="red")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 3️⃣ Bar chart — comparer des catégories

**Quand l'utiliser ?** Pour comparer des valeurs entre des catégories discrètes.

```python
#| fig-cap: "CA par catégorie de produit"

# Préparer
ca_cat = df.groupby("categorie")["montant_total"].sum().sort_values(ascending=True)

fig, ax = plt.subplots(figsize=(10, 5))

bars = ax.barh(ca_cat.index, ca_cat.values, color="steelblue", edgecolor="black", alpha=0.8)

ax.set_xlabel("CA (€)")
ax.set_title("Chiffre d'affaires par catégorie")

# Annoter chaque barre avec sa valeur
for bar, value in zip(bars, ca_cat.values):
    ax.text(bar.get_width() + 500, bar.get_y() + bar.get_height()/2,
            f"{value:,.0f} €", va="center", fontsize=10)

ax.grid(True, alpha=0.3, axis="x")
plt.tight_layout()
plt.show()
```

> **💡 Astuce**
>
## 🧠 Astuce
Les barres **horizontales** (`barh`) sont souvent plus lisibles que les verticales (`bar`) quand les labels sont longs. Toujours trier les barres par taille pour faciliter la lecture.


## 4️⃣ Histogramme — distribution d'une variable numérique

**Quand l'utiliser ?** Pour voir la **forme** de la distribution d'une variable.

```python
#| fig-cap: "Distribution des montants de transaction"

fig, ax = plt.subplots(figsize=(10, 5))

ax.hist(df["montant_total"], bins=50, color="steelblue", 
        edgecolor="black", alpha=0.7)

ax.axvline(df["montant_total"].mean(), color="red", linestyle="--", 
           linewidth=2, label=f"Moyenne : {df['montant_total'].mean():.1f} €")
ax.axvline(df["montant_total"].median(), color="green", linestyle="--", 
           linewidth=2, label=f"Médiane : {df['montant_total'].median():.1f} €")

ax.set_xlabel("Montant total (€)")
ax.set_ylabel("Nombre de transactions")
ax.set_title("Distribution des montants de transaction")
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 5️⃣ Boxplot — comparer les distributions

**Quand l'utiliser ?** Pour comparer la distribution d'une variable numérique selon une catégorie.

```python
#| fig-cap: "Distribution des montants par catégorie"

fig, ax = plt.subplots(figsize=(10, 5))

# Préparer les données par catégorie
cats = df["categorie"].unique()
donnees_par_cat = [df[df["categorie"] == c]["montant_total"].values for c in cats]

bp = ax.boxplot(donnees_par_cat, tick_labels=cats, patch_artist=True)

# Colorer les boîtes
colors = ["#4C72B0", "#DD8452", "#55A868", "#C44E52", "#8172B3"]
for patch, color in zip(bp["boxes"], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax.set_ylabel("Montant (€)")
ax.set_title("Distribution des montants par catégorie")
ax.grid(True, alpha=0.3, axis="y")
plt.tight_layout()
plt.show()
```

Dans un boxplot on lit :
- La **boîte** : Q1 → Q3 (50% des données)
- La **ligne** : médiane
- Les **moustaches** : étendue "normale"
- Les **points** : outliers

## 6️⃣ Heatmap — visualiser une matrice

**Quand l'utiliser ?** Pour visualiser une matrice 2D : corrélations, tableau croisé, similarités.

```python
#| fig-cap: "CA par catégorie × région (tableau croisé)"

# Préparer la matrice avec pivot_table
pivot = df.pivot_table(
    values="montant_total",
    index="categorie",
    columns="region",
    aggfunc="sum",
    fill_value=0
)

fig, ax = plt.subplots(figsize=(12, 5))

sns.heatmap(pivot, annot=True, fmt=",.0f", cmap="YlOrRd", 
            cbar_kws={"label": "CA (€)"}, ax=ax)

ax.set_title("CA par catégorie et région")
plt.tight_layout()
plt.show()
```

# Seaborn : la data viz statistique élégante

**Seaborn** est une surcouche de Matplotlib qui **simplifie énormément** les graphiques statistiques. Il comprend nativement Pandas.

## 🎨 Philosophie : déclarer, pas dessiner

Au lieu de préparer les données puis les tracer, tu **décris** ce que tu veux. Seaborn s'occupe du reste.

```python
#| fig-cap: "Seaborn : scatter avec semantic mapping en une ligne"

fig, ax = plt.subplots(figsize=(10, 5))
sns.scatterplot(data=df, x="quantite", y="montant_total", 
                hue="categorie", alpha=0.6, ax=ax)
ax.set_title("Montant vs quantité, coloré par catégorie")
plt.tight_layout()
plt.show()
```

En **une ligne** : le scatter plot avec coloration par catégorie (hue) **et** la légende qui va bien.

## 📊 Les 5 fonctions Seaborn à connaître

### `sns.histplot` — histogramme avancé

```python
#| fig-cap: "Histogrammes par groupe"

fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(data=df, x="montant_total", hue="categorie", 
             multiple="stack", bins=30, ax=ax)
ax.set_title("Distribution des montants par catégorie (empilés)")
plt.tight_layout()
plt.show()
```

### `sns.boxplot` — boxplots par groupe

```python
#| fig-cap: "Boxplot avec Seaborn"

fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(data=df, x="categorie", y="montant_total", ax=ax)
ax.set_title("Montants par catégorie")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

### `sns.violinplot` — un boxplot avec la densité

```python
#| fig-cap: "Violinplot : boxplot + densité"

fig, ax = plt.subplots(figsize=(10, 5))
sns.violinplot(data=df, x="categorie", y="montant_total", ax=ax)
ax.set_title("Distribution et densité des montants par catégorie")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

Le violinplot est un boxplot **où la largeur montre la densité** à chaque niveau. Très utile pour voir des distributions multimodales.

### `sns.countplot` — compter les occurrences

```python
#| fig-cap: "Compter les modalités d'une variable catégorielle"

fig, ax = plt.subplots(figsize=(10, 4))
sns.countplot(data=df, y="region", 
              order=df["region"].value_counts().index, ax=ax)
ax.set_title("Nombre de transactions par région")
ax.set_xlabel("Nombre de transactions")
plt.tight_layout()
plt.show()
```

### `sns.heatmap` — matrice colorée

Déjà vu plus haut. Incontournable pour visualiser des corrélations et des tableaux croisés.

## 🌟 `sns.pairplot` : l'exploration massive

Pour **explorer toutes les relations** entre variables numériques d'un DataFrame d'un seul coup.

```python
#| fig-cap: "Pairplot : toutes les relations croisées"

# Un sous-ensemble pour que ce soit rapide
df_num = df[["quantite", "prix_unitaire", "montant_total", "categorie"]].sample(300, random_state=42)

g = sns.pairplot(df_num, hue="categorie", height=2.5, 
                  diag_kind="hist", plot_kws={"alpha": 0.5})
g.fig.suptitle("Relations croisées entre variables", y=1.02, fontsize=13)
plt.tight_layout()
plt.show()
```

En **une commande**, tu obtiens :
- Les scatter plots de toutes les paires
- Les histogrammes sur la diagonale
- Les couleurs par catégorie

C'est l'outil d'**exploration initiale** par excellence.

# Personnalisation : de "bof" à "pro"

Les bases des graphiques, c'est bien. La différence entre un graphique amateur et un graphique pro tient à quelques détails.

## 🎨 Les 5 règles d'or

1. **Titres et labels complets** : un graphique sans titre ne sert à rien
2. **Unités** : mettre les unités dans les labels (`€`, `kg`, `°C`...)
3. **Taille de texte lisible** : `fontsize=12` minimum
4. **Pas trop de couleurs** : max 5-6 différentes pour lisibilité
5. **Annotations** : pointer les valeurs clés directement sur le graphique

## 🖼️ Styles Seaborn

```python
#| fig-cap: "Les 5 styles Seaborn disponibles"

styles = ["whitegrid", "white", "darkgrid", "dark", "ticks"]

fig, axes = plt.subplots(1, 5, figsize=(15, 3))
x = np.linspace(0, 10, 100)

for ax, style in zip(axes, styles):
    with sns.axes_style(style):
        ax.clear()
        ax.plot(x, np.sin(x))
        ax.set_title(style)

plt.tight_layout()
plt.show()

# Réappliquer le style par défaut
sns.set_theme(style="whitegrid")
```

## 🎨 Palettes de couleurs

Seaborn a des **palettes adaptées à chaque usage** :

```python
#| fig-cap: "Les 3 grandes familles de palettes"

fig, axes = plt.subplots(3, 1, figsize=(10, 4))

# Qualitative (pour des catégories sans ordre)
sns.palplot(sns.color_palette("Set2", 8))
axes[0].set_title("Qualitative (Set2) — pour catégories sans ordre")

# Séquentielle (pour valeurs ordonnées)
sns.palplot(sns.color_palette("Blues", 8))
axes[1].set_title("Séquentielle (Blues) — pour valeurs ordonnées")

# Divergente (pour valeurs centrées sur 0)
sns.palplot(sns.color_palette("RdBu", 8))
axes[2].set_title("Divergente (RdBu) — pour valeurs centrées sur 0")

plt.tight_layout()
plt.show()
```

> **💡 Astuce**
>
## 🧠 Règle simple
- Catégorielle (régions, types...) → palettes qualitatives : `Set1`, `Set2`, `tab10`
- Ordonnée (classements, intensités) → palettes séquentielles : `Blues`, `YlOrRd`
- Différences (hausse/baisse) → palettes divergentes : `RdBu`, `coolwarm`

2. **Toujours titre + labels + unités**. Un graphique nu n'a aucune valeur.
3. **Seaborn pour les graphiques statistiques**, Matplotlib pour le contrôle fin.
4. **Palettes pertinentes** : qualitative / séquentielle / divergente selon la donnée.
5. **Annotations directes** mieux qu'une légende séparée quand c'est possible.
