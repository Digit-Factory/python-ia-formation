```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import joblib
import tempfile
from pathlib import Path

# Charger et split
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42, stratify=data.target
)

# Entraîner
modele = RandomForestClassifier(n_estimators=100, random_state=42)
modele.fit(X_train, y_train)

# Évaluer
print(f"Accuracy test : {modele.score(X_test, y_test):.3f}")

# Sauvegarder (chemin portable : fonctionne sur Linux, macOS et Windows)
MODELS_DIR = Path(tempfile.gettempdir()) / "models"
MODELS_DIR.mkdir(parents=True, exist_ok=True)
joblib.dump(modele, MODELS_DIR / "iris_model.joblib")
joblib.dump(list(data.target_names), MODELS_DIR / "iris_labels.joblib")
print(f"✅ Modèle sauvegardé dans {MODELS_DIR}")
```

```python
from pydantic import BaseModel, Field
from typing import Literal

class IrisFeatures(BaseModel):
    """Features pour prédire une espèce d'iris."""
    sepal_length: float = Field(..., gt=0, lt=20, description="Longueur du sépale (cm)")
    sepal_width: float = Field(..., gt=0, lt=20, description="Largeur du sépale (cm)")
    petal_length: float = Field(..., gt=0, lt=20, description="Longueur du pétale (cm)")
    petal_width: float = Field(..., gt=0, lt=20, description="Largeur du pétale (cm)")

class IrisPrediction(BaseModel):
    """Résultat de prédiction."""
    species: Literal["setosa", "versicolor", "virginica"]
    confidence: float = Field(..., ge=0, le=1, description="Probabilité de la classe")
    model_version: str

# Test : valider une instance
fleur = IrisFeatures(sepal_length=5.1, sepal_width=3.5, petal_length=1.4, petal_width=0.2)
print(fleur.model_dump())
```

```python
from fastapi import FastAPI, HTTPException
import joblib
import numpy as np

# Nouvelle app
app = FastAPI(title="Iris Classifier", version="1.0.0")

# Charger le modèle au démarrage (une fois)
modele_prod = joblib.load(MODELS_DIR / "iris_model.joblib")
labels_prod = joblib.load(MODELS_DIR / "iris_labels.joblib")
MODEL_VERSION = "rf-v1.0.0"

@app.get("/")
def root():
    return {"service": "iris-classifier", "version": MODEL_VERSION}

@app.post("/predict", response_model=IrisPrediction)
def predict(features: IrisFeatures):
    """Prédit l'espèce d'iris à partir de ses mesures."""
    # 1. Convertir en array
    X = np.array([[
        features.sepal_length, features.sepal_width,
        features.petal_length, features.petal_width,
    ]])
    
    # 2. Prédire
    pred_idx = int(modele_prod.predict(X)[0])
    probas = modele_prod.predict_proba(X)[0]
    
    # 3. Retourner
    return IrisPrediction(
        species=labels_prod[pred_idx],
        confidence=float(probas[pred_idx]),
        model_version=MODEL_VERSION,
    )

print("✅ Endpoint /predict défini")
```

```python
from typing import List
from pydantic import Field as PydField

class IrisBatch(BaseModel):
    items: List[IrisFeatures] = PydField(..., max_length=100)

class BatchPrediction(BaseModel):
    predictions: List[IrisPrediction]
    count: int

@app.post("/predict-batch", response_model=BatchPrediction)
def predict_batch(batch: IrisBatch):
    """Prédit plusieurs iris en une requête (plus efficace)."""
    # Construire matrice X (N, 4)
    X = np.array([
        [item.sepal_length, item.sepal_width, item.petal_length, item.petal_width]
        for item in batch.items
    ])
    
    # Prédire en batch
    pred_idx = modele_prod.predict(X).astype(int)
    probas = modele_prod.predict_proba(X)
    
    # Construire les prédictions
    predictions = [
        IrisPrediction(
            species=labels_prod[int(idx)],
            confidence=float(probas[i, idx]),
            model_version=MODEL_VERSION,
        )
        for i, idx in enumerate(pred_idx)
    ]
    
    return BatchPrediction(predictions=predictions, count=len(predictions))

# Tester
client = TestClient(app)
response = client.post("/predict-batch", json={
    "items": [
        {"sepal_length": 5.1, "sepal_width": 3.5, "petal_length": 1.4, "petal_width": 0.2},
        {"sepal_length": 6.5, "sepal_width": 3.0, "petal_length": 5.5, "petal_width": 2.0},
        {"sepal_length": 5.9, "sepal_width": 3.0, "petal_length": 4.2, "petal_width": 1.5},
    ]
})
print(f"Status : {response.status_code}")
import json as jsonlib
print(jsonlib.dumps(response.json(), indent=2))
```

```python
from fastapi import FastAPI, HTTPException
from fastapi.testclient import TestClient
from pydantic import BaseModel, Field
from typing import Literal

app_techvolt = FastAPI(title="TechVolt API", version="1.0.0")

# --- Base de connaissance ---
DB_ERREURS = {
    "E01": {"urgence": "NORMALE", "action": "Laisser refroidir 30min"},
    "E02": {"urgence": "HAUTE", "action": "Ne pas utiliser, appeler électricien"},
    "E03": {"urgence": "NORMALE", "action": "Vérifier wifi et redémarrer"},
    "E04": {"urgence": "URGENTE", "action": "Débrancher, couper disjoncteur, appeler 0800 000 000"},
    "E05": {"urgence": "NORMALE", "action": "Vérifier câble et connecteur"},
}

MOTS_POSITIFS = {"super", "excellent", "parfait", "adore", "rapide", "professionnel"}
MOTS_NEGATIFS = {"nul", "cassé", "panne", "bug", "déçu", "inadmissible", "lent"}

# --- Schémas Pydantic ---
class ErreurInput(BaseModel):
    code: str = Field(..., pattern=r"^E\d{2}$")

class ErreurOutput(BaseModel):
    code: str
    urgence: Literal["NORMALE", "HAUTE", "URGENTE"]
    action: str

class AvisInput(BaseModel):
    texte: str = Field(..., min_length=5)

class AvisOutput(BaseModel):
    sentiment: Literal["positif", "neutre", "negatif"]
    confidence: float = Field(..., ge=0, le=1)

# --- Endpoints ---
@app_techvolt.get("/health")
def health():
    return {"status": "ok", "version": "1.0.0"}

@app_techvolt.post("/diagnostic-erreur", response_model=ErreurOutput)
def diagnostic(err: ErreurInput):
    info = DB_ERREURS.get(err.code)
    if info is None:
        raise HTTPException(status_code=404, detail=f"Code {err.code} inconnu")
    return ErreurOutput(code=err.code, **info)

@app_techvolt.post("/classifier-avis", response_model=AvisOutput)
def classifier(avis: AvisInput):
    mots = avis.texte.lower().split()
    n_pos = sum(1 for m in mots if m in MOTS_POSITIFS)
    n_neg = sum(1 for m in mots if m in MOTS_NEGATIFS)
    total = max(n_pos + n_neg, 1)
    
    if n_pos > n_neg:
        return AvisOutput(sentiment="positif", confidence=n_pos / total)
    elif n_neg > n_pos:
        return AvisOutput(sentiment="negatif", confidence=n_neg / total)
    else:
        return AvisOutput(sentiment="neutre", confidence=0.5)

# --- Tests ---
c = TestClient(app_techvolt)

# Test 1 : health
r = c.get("/health")
assert r.status_code == 200 and r.json()["status"] == "ok"
print("✅ test_health PASSED")

# Test 2 : diagnostic E04
r = c.post("/diagnostic-erreur", json={"code": "E04"})
assert r.status_code == 200 and r.json()["urgence"] == "URGENTE"
print("✅ test_diagnostic_E04 PASSED")

# Test 3 : diagnostic code invalide (format)
r = c.post("/diagnostic-erreur", json={"code": "XYZ"})
assert r.status_code == 422  # validation Pydantic
print("✅ test_diagnostic_code_invalide_format PASSED")

# Test 4 : diagnostic code inconnu
r = c.post("/diagnostic-erreur", json={"code": "E99"})
assert r.status_code == 404
print("✅ test_diagnostic_code_inconnu PASSED")

# Test 5 : avis positif
r = c.post("/classifier-avis", json={"texte": "Installation rapide et technicien professionnel, parfait !"})
assert r.json()["sentiment"] == "positif"
print("✅ test_avis_positif PASSED")

# Test 6 : avis négatif
r = c.post("/classifier-avis", json={"texte": "Ma borne est cassée, inadmissible de la part du SAV"})
assert r.json()["sentiment"] == "negatif"
print("✅ test_avis_negatif PASSED")

# Test 7 : avis trop court
r = c.post("/classifier-avis", json={"texte": "ok"})
assert r.status_code == 422
print("✅ test_avis_trop_court PASSED")

print("\n🎉 Tous les tests passent !")
```
