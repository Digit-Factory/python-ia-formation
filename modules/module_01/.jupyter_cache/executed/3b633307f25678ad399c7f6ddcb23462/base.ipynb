```python
def derivee_numerique(f, x, h=1e-5):
    """Calcule la dérivée de f en x par différences finies."""
    return (f(x + h) - f(x)) / h

# Testons sur f(x) = x² (dérivée connue : f'(x) = 2x)
f = lambda x: x**2

print(f"f'(3) calculée   : {derivee_numerique(f, 3):.4f}")
print(f"f'(3) théorique  : {2*3}")  # Pour x=3, 2x = 6
print()
print(f"f'(-2) calculée  : {derivee_numerique(f, -2):.4f}")
print(f"f'(-2) théorique : {2*(-2)}")
```

```python
#| label: fig-derivee-pente
#| fig-cap: La dérivée d'une fonction = la pente de la tangente
# Fonction test
f = lambda x: x**2

# Plage de tracé
x = np.linspace(-3, 3, 100)
y = f(x)

# Point où on calcule la dérivée
x0 = 1.5
y0 = f(x0)
pente = derivee_numerique(f, x0)

# Tangente : y = pente * (x - x0) + y0
x_tang = np.linspace(x0 - 1, x0 + 1, 10)
y_tang = pente * (x_tang - x0) + y0

fig, ax = plt.subplots()
ax.plot(x, y, 'b-', linewidth=2, label='f(x) = x²')
ax.plot(x_tang, y_tang, 'r-', linewidth=2, label=f'Tangente (pente = {pente:.2f})')
ax.plot(x0, y0, 'ko', markersize=10)
ax.annotate(f'Point ({x0}, {y0:.2f})', xy=(x0, y0), xytext=(x0+0.3, y0-0.5),
            fontsize=11, arrowprops=dict(arrowstyle='->'))
ax.axhline(y=0, color='gray', linewidth=0.5)
ax.axvline(x=0, color='gray', linewidth=0.5)
ax.grid(True, alpha=0.3)
ax.legend()
ax.set_title(f"Dérivée au point x={x0} : pente de la tangente = {pente:.2f}")
plt.show()
```

```python
def gradient_numerique(f, x, y, h=1e-5):
    """Calcule le gradient de f(x, y) au point (x, y)."""
    df_dx = (f(x + h, y) - f(x, y)) / h
    df_dy = (f(x, y + h) - f(x, y)) / h
    return np.array([df_dx, df_dy])

# Test sur f(x, y) = x² + y²
f = lambda x, y: x**2 + y**2

# Au point (1, 2) : gradient théorique = (2*1, 2*2) = (2, 4)
grad = gradient_numerique(f, 1, 2)
print(f"Gradient au point (1, 2) : {grad}")
print(f"Théorique                 : [2, 4]")
```

```python
def descente_gradient_1d(f, df, x_depart, learning_rate, n_iterations):
    """
    Descente de gradient en 1D.
    Retourne l'historique des x pour pouvoir visualiser.
    """
    x = x_depart
    historique = [x]
    
    for _ in range(n_iterations):
        gradient = df(x)
        x = x - learning_rate * gradient
        historique.append(x)
    
    return x, historique

# Notre fonction et sa dérivée (calculée à la main ici, mais on pourrait la faire numériquement)
f = lambda x: x**2 - 4*x + 7
df = lambda x: 2*x - 4

# Lançons la descente
x_final, historique = descente_gradient_1d(
    f=f, df=df, 
    x_depart=10,       # on part loin du minimum
    learning_rate=0.1, # pas raisonnable
    n_iterations=30
)

print(f"Point de départ : 10")
print(f"Point final     : {x_final:.4f}")
print(f"Valeur théorique du minimum : 2")
print(f"Nombre d'itérations         : {len(historique)}")
```

```python
#| label: fig-descente-trajectoire
#| fig-cap: La descente de gradient en action
x_range = np.linspace(-1, 11, 100)
y_range = f(x_range)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x_range, y_range, 'b-', linewidth=2, label='f(x) = x² - 4x + 7')

# Tracer chaque étape de la descente
hist_x = np.array(historique)
hist_y = f(hist_x)
ax.plot(hist_x, hist_y, 'ro-', markersize=6, alpha=0.7, label='Trajectoire')
ax.plot(hist_x[0], hist_y[0], 'go', markersize=12, label='Départ')
ax.plot(hist_x[-1], hist_y[-1], 'y*', markersize=20, label='Arrivée')

ax.axvline(x=2, color='gray', linestyle='--', alpha=0.5, label='Minimum théorique (x=2)')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title("Descente de gradient : la balle dévale la cuvette")
ax.legend()
ax.grid(True, alpha=0.3)
plt.show()
```

```python
# Générer des données bruitées (taille en cm, poids en kg)
np.random.seed(42)
taille = np.random.uniform(150, 200, 20)
poids_vrai = 0.9 * taille - 80 + np.random.normal(0, 5, 20)

plt.figure(figsize=(9, 6))
plt.scatter(taille, poids_vrai, s=50, alpha=0.7, edgecolors='k')
plt.xlabel('Taille (cm)')
plt.ylabel('Poids (kg)')
plt.title('Relation taille / poids (20 personnes)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Normalisation (centrage-réduction)
taille_mean = taille.mean()
taille_std = taille.std()
poids_mean = poids_vrai.mean()
poids_std = poids_vrai.std()

taille_norm = (taille - taille_mean) / taille_std
poids_norm = (poids_vrai - poids_mean) / poids_std

print(f"Taille avant : moyenne = {taille.mean():.1f}, std = {taille.std():.1f}")
print(f"Taille après : moyenne = {taille_norm.mean():.3f}, std = {taille_norm.std():.3f}")
```

```python
def mse(a, b, x, y):
    """Erreur quadratique moyenne."""
    predictions = a * x + b
    return np.mean((predictions - y)**2)

def gradient_mse(a, b, x, y):
    """Gradient de la MSE par rapport à a et b."""
    n = len(x)
    predictions = a * x + b
    errors = predictions - y
    grad_a = (2/n) * np.sum(errors * x)
    grad_b = (2/n) * np.sum(errors)
    return grad_a, grad_b

def regression_par_descente(x, y, learning_rate=0.1, n_iterations=1000):
    """Entraîne une régression linéaire par descente de gradient."""
    a, b = 0.0, 0.0
    historique = []
    
    for i in range(n_iterations):
        grad_a, grad_b = gradient_mse(a, b, x, y)
        a = a - learning_rate * grad_a
        b = b - learning_rate * grad_b
        
        if i % 100 == 0 or i == n_iterations - 1:
            erreur = mse(a, b, x, y)
            historique.append((i, a, b, erreur))
    
    return a, b, historique

# Entraînement sur les données normalisées !
a_norm, b_norm, hist = regression_par_descente(
    taille_norm, poids_norm, 
    learning_rate=0.1,
    n_iterations=1000
)

print(f"Sur les données normalisées : a = {a_norm:.4f}, b = {b_norm:.4f}")
print()
print("Progression de l'erreur :")
for i, a, b, err in hist[:3]:
    print(f"  Itération {i:4d} : a = {a:7.4f}, b = {b:7.4f}, MSE = {err:.4f}")
print("  ...")
for i, a, b, err in hist[-2:]:
    print(f"  Itération {i:4d} : a = {a:7.4f}, b = {b:7.4f}, MSE = {err:.4f}")
```

```python
# 1. Les données
np.random.seed(0)
surfaces = np.random.uniform(20, 150, 50)
occupants = np.random.randint(1, 6, 50).astype(float)
conso = 2.5 * surfaces + 15 * occupants + 10 + np.random.normal(0, 20, 50)

# 2. Normalisation
surfaces_norm = (surfaces - surfaces.mean()) / surfaces.std()
occupants_norm = (occupants - occupants.mean()) / occupants.std()
conso_norm = (conso - conso.mean()) / conso.std()

# 3. MSE à 3 paramètres
def mse_multi(a, b, c, x1, x2, y):
    predictions = a * x1 + b * x2 + c
    return np.mean((predictions - y)**2)

# 4. Gradient
def gradient_mse_multi(a, b, c, x1, x2, y):
    n = len(y)
    predictions = a * x1 + b * x2 + c
    errors = predictions - y
    grad_a = (2/n) * np.sum(errors * x1)
    grad_b = (2/n) * np.sum(errors * x2)
    grad_c = (2/n) * np.sum(errors)
    return grad_a, grad_b, grad_c

# 5. Descente sur les données normalisées
a, b, c = 0.0, 0.0, 0.0
lr = 0.1
n_iter = 1000
erreurs = []

for i in range(n_iter):
    grad_a, grad_b, grad_c = gradient_mse_multi(
        a, b, c, surfaces_norm, occupants_norm, conso_norm
    )
    a -= lr * grad_a
    b -= lr * grad_b
    c -= lr * grad_c
    erreurs.append(mse_multi(a, b, c, surfaces_norm, occupants_norm, conso_norm))

# 6. Résultats
print(f"Coefficients sur données normalisées :")
print(f"  a = {a:.4f}  (influence de la surface)")
print(f"  b = {b:.4f}  (influence du nb d'occupants)")
print(f"  c = {c:.4f}  (terme constant)")
print(f"  MSE finale = {erreurs[-1]:.6f}")
print()
print(f"Les valeurs sont interprétables :")
print(f"  → a > b signifie que la surface a plus d'influence que le nb d'occupants")
print(f"  → c ≈ 0 car on a centré les données")

# 7. Courbe d'erreur
plt.figure(figsize=(10, 5))
plt.plot(erreurs)
plt.xlabel('Itération')
plt.ylabel('MSE (sur données normalisées)')
plt.title('Convergence de la descente de gradient')
plt.yscale('log')
plt.grid(True, alpha=0.3)
plt.show()
```
