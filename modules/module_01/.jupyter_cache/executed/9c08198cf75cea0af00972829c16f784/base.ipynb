```python
taille_avant = len(df)

# Fonction utilitaire
def detecter_outliers_iqr(serie):
    q1 = serie.quantile(0.25)
    q3 = serie.quantile(0.75)
    iqr = q3 - q1
    borne_basse = q1 - 1.5 * iqr
    borne_haute = q3 + 1.5 * iqr
    return (serie < borne_basse) | (serie > borne_haute)

# Détection sur surface et prix
outliers_surface = detecter_outliers_iqr(df['surface_m2'])
outliers_prix = detecter_outliers_iqr(df['prix_eur'])

print(f"Outliers surface_m2 : {outliers_surface.sum()}")
print(f"Outliers prix_eur   : {outliers_prix.sum()}")

# Supprimer les lignes outliers (sur l'une OU l'autre variable)
df = df[~(outliers_surface | outliers_prix)].reset_index(drop=True)

print(f"\nLignes supprimées : {taille_avant - len(df)}")
print(f"Taille finale : {len(df)} lignes")
```

```python
def gradient_mse(X, y, w, b):
    """Retourne (grad_w, grad_b) : gradients de la MSE."""
    n = len(y)
    predictions = predire(X, w, b)
    erreurs = predictions - y
    
    grad_w = (2/n) * (X.T @ erreurs)  # vecteur de taille (d,)
    grad_b = (2/n) * np.sum(erreurs)  # scalaire
    
    return grad_w, grad_b
```

```python
def entrainer(X, y, learning_rate=0.05, n_iter=2000):
    """Entraîne une régression linéaire par descente de gradient."""
    d = X.shape[1]
    w = np.zeros(d)
    b = 0.0
    historique = []
    
    for i in range(n_iter):
        grad_w, grad_b = gradient_mse(X, y, w, b)
        w = w - learning_rate * grad_w
        b = b - learning_rate * grad_b
        historique.append(mse(X, y, w, b))
    
    return w, b, historique

# Entraînement
w, b, historique = entrainer(X_train_norm, y_train_norm, learning_rate=0.05, n_iter=2000)

print(f"MSE initiale : {historique[0]:.4f}")
print(f"MSE finale   : {historique[-1]:.4f}")
print(f"Division par : {historique[0] / historique[-1]:.1f}x")
```

```python
def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred)**2))

def r_squared(y_true, y_pred):
    ss_res = np.sum((y_true - y_pred)**2)
    ss_tot = np.sum((y_true - y_true.mean())**2)
    return 1 - (ss_res / ss_tot)

mae_test = mae(y_test, y_pred)
rmse_test = rmse(y_test, y_pred)
r2_test = r_squared(y_test, y_pred)

print("=== Performances sur le test set ===")
print(f"MAE  : {mae_test:,.0f} €       (erreur absolue moyenne)")
print(f"RMSE : {rmse_test:,.0f} €      (racine de l'erreur quadratique moyenne)")
print(f"R²   : {r2_test:.4f}            (plus proche de 1, mieux)")
print(f"\nPrix moyen réel : {y_test.mean():,.0f} €")
print(f"Erreur relative moyenne : {mae_test / y_test.mean():.1%}")
```
