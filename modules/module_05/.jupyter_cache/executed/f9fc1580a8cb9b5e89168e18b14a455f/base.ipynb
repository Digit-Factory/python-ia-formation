```python
# Garder id_client à part
ids = df["id_client"].copy()
X = df.drop(columns="id_client")

# Imputation médiane
imputer = SimpleImputer(strategy="median")
X_imp = pd.DataFrame(
    imputer.fit_transform(X),
    columns=X.columns,
    index=X.index
)

print(f"NaN restants : {X_imp.isna().sum().sum()}")
print(f"Shape : {X_imp.shape}")
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imp)

# Vérification
print(f"Moyennes (doivent être ~0) : {X_scaled.mean(axis=0).round(3)}")
print(f"Écarts-types (doivent être ~1) : {X_scaled.std(axis=0).round(3)}")
```

```python
pca_full = PCA()
pca_full.fit(X_scaled)

cumul = np.cumsum(pca_full.explained_variance_ratio_)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(range(1, len(cumul) + 1), cumul * 100, "o-", linewidth=2, markersize=8)
for seuil, couleur in [(80, "green"), (90, "orange"), (95, "red")]:
    n_c = np.argmax(cumul >= seuil/100) + 1
    ax.axhline(seuil, color=couleur, linestyle="--", alpha=0.4)
    ax.axvline(n_c, color=couleur, linestyle="--", alpha=0.4)
    ax.text(n_c + 0.1, seuil - 2, f"{seuil}% → {n_c} comp.", color=couleur, fontsize=10)
ax.set_xlabel("Nombre de composantes")
ax.set_ylabel("Variance cumulée (%)")
ax.set_title("Variance expliquée cumulée")
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
reducer = umap.UMAP(n_neighbors=30, min_dist=0.1, random_state=42)
X_umap = reducer.fit_transform(X_scaled)

fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(X_umap[:, 0], X_umap[:, 1], s=10, alpha=0.5, color="steelblue", edgecolor="white", linewidth=0.2)
ax.set_title("UMAP 2D — 5000 clients")
ax.set_xlabel("UMAP 1"); ax.set_ylabel("UMAP 2")
plt.tight_layout()
plt.show()
```

```python
inerties = []
silhouettes = []
k_range = range(2, 11)

for k in k_range:
    km = KMeans(n_clusters=k, n_init=10, random_state=42)
    labels = km.fit_predict(X_scaled)
    inerties.append(km.inertia_)
    silhouettes.append(silhouette_score(X_scaled, labels))

# Trace
fig, axes = plt.subplots(1, 2, figsize=(14, 4))
axes[0].plot(k_range, inerties, "o-", linewidth=2, markersize=8, color="steelblue")
axes[0].set_xlabel("k"); axes[0].set_ylabel("Inertie")
axes[0].set_title("Méthode du coude")
axes[0].grid(True, alpha=0.3)

axes[1].plot(k_range, silhouettes, "o-", linewidth=2, markersize=8, color="coral")
axes[1].set_xlabel("k"); axes[1].set_ylabel("Silhouette")
axes[1].set_title("Coefficient de silhouette")
axes[1].grid(True, alpha=0.3)

k_opt = list(k_range)[np.argmax(silhouettes)]
axes[1].axvline(k_opt, color="green", linestyle="--", alpha=0.5,
                 label=f"Max silhouette : k={k_opt}")
axes[1].legend()

plt.tight_layout()
plt.show()

print(f"k optimal selon silhouette : {k_opt}")
```

```python
iso = IsolationForest(contamination=0.01, random_state=42)
predictions = iso.fit_predict(X_scaled)
df["anomalie"] = (predictions == -1).astype(int)
df["score_anomalie"] = iso.decision_function(X_scaled)

print(f"Clients flaggés comme anomalies : {df['anomalie'].sum()}")
```

```python
# Visualisation sur UMAP
fig, ax = plt.subplots(figsize=(10, 8))
ax.scatter(X_umap[df["anomalie"] == 0, 0], X_umap[df["anomalie"] == 0, 1],
           s=10, alpha=0.4, color="steelblue", label="Normal")
ax.scatter(X_umap[df["anomalie"] == 1, 0], X_umap[df["anomalie"] == 1, 1],
           s=80, marker="X", color="red", edgecolor="black", linewidth=1,
           label=f"Anomalie ({df['anomalie'].sum()})")
ax.set_title("Détection d'anomalies — clients suspects")
ax.legend(markerscale=1)
plt.tight_layout()
plt.show()
```
