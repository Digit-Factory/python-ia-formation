```python
# Chargement
df_ventes = pd.read_csv("ressources_tp/ventes_ecommerce.csv", parse_dates=["date"])

# 1. Audit typologique
print("=== Audit typologique ===\n")

audit = pd.DataFrame({
    "colonne": df_ventes.columns,
    "dtype_actuel": df_ventes.dtypes.values
})
print(audit)
```

```python
# 2. Conversions
for col in ["nom_produit", "categorie", "region", "mode_paiement", "statut"]:
    df_ventes[col] = df_ventes[col].astype("category")

# 3. Vérification
print(df_ventes.info())
```

```python
# 4. Graphiques adaptés
fig, axes = plt.subplots(2, 2, figsize=(15, 8))

# Quantitative continue : montant_total
axes[0, 0].hist(df_ventes["montant_total"], bins=40, color="steelblue", 
                edgecolor="black", alpha=0.7)
axes[0, 0].set_title("Histogramme : montant_total (quanti continue)")
axes[0, 0].set_xlabel("Montant (€)")

# Qualitative nominale : categorie
df_ventes["categorie"].value_counts().plot(
    kind="bar", ax=axes[0, 1], color="coral", edgecolor="black"
)
axes[0, 1].set_title("Bar chart : categorie (quali nominale)")
axes[0, 1].tick_params(axis="x", rotation=45)

# Qualitative nominale ordonnée (on traite statut comme nominal ici)
df_ventes["statut"].value_counts().plot(
    kind="bar", ax=axes[1, 0], color="mediumseagreen", edgecolor="black"
)
axes[1, 0].set_title("Bar chart : statut (quali nominale)")
axes[1, 0].tick_params(axis="x", rotation=45)

# Temporelle : évolution mensuelle
ca_mensuel = df_ventes.groupby(df_ventes["date"].dt.to_period("M"))["montant_total"].sum()
axes[1, 1].plot(ca_mensuel.index.astype(str), ca_mensuel.values, 
                marker="o", color="purple")
axes[1, 1].set_title("Line plot : CA mensuel (temporelle)")
axes[1, 1].tick_params(axis="x", rotation=45)

for ax in axes.flat:
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
