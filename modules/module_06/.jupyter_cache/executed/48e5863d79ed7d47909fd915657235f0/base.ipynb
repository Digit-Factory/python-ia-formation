```python
# Recharger le modèle (pour partir de zéro)
modele = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)

# Tout geler
for param in modele.parameters():
    param.requires_grad = False

# Vérifier qu'aucun paramètre n'est entraîné
n_params_entraines = sum(p.numel() for p in modele.parameters() if p.requires_grad)
print(f"Paramètres entraînables après gel : {n_params_entraines}")
```

```python
# Récupérer la taille d'entrée de fc
n_features_in = modele.fc.in_features
print(f"Features en entrée de fc : {n_features_in}")

# Remplacer fc pour 10 classes (exemple : CIFAR-10)
modele.fc = nn.Linear(n_features_in, 10)

# Maintenant, SEULE fc est entraînable
n_params_entraines = sum(p.numel() for p in modele.parameters() if p.requires_grad)
print(f"Paramètres entraînables : {n_params_entraines}")
print(f"   = {n_features_in} × 10 + 10 = 5130")
```

```python
# Transformations IMPORTANTES : il faut les MÊMES qu'à l'entraînement ImageNet
# (sinon le modèle "voit" des choses différentes)
transform_train = transforms.Compose([
    transforms.Resize(224),              # ResNet veut 224×224
    transforms.RandomHorizontalFlip(),    # data augmentation
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize(                 # stats ImageNet (OBLIGATOIRE)
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

transform_test = transforms.Compose([
    transforms.Resize(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# CIFAR-10 : 10 classes d'objets en couleur (32×32 → resize en 224×224)
# avion, auto, oiseau, chat, cerf, chien, grenouille, cheval, bateau, camion
print("✅ Transforms définis (224×224, stats ImageNet)")
```

```python
#| eval: false
# Téléchargement CIFAR-10 (~170 MB, une seule fois)
train_full = torchvision.datasets.CIFAR10(
    root=os.path.join(DATA_ROOT, "cifar"), train=True, download=True, transform=transform_train
)
test_set = torchvision.datasets.CIFAR10(
    root=os.path.join(DATA_ROOT, "cifar"), train=False, download=True, transform=transform_test
)

# Sous-échantillonnage : seulement 500 images train pour démontrer l'efficacité
indices = np.random.choice(len(train_full), 500, replace=False).tolist()
train_small = Subset(train_full, indices)

train_loader = DataLoader(train_small, batch_size=32, shuffle=True)
test_loader = DataLoader(test_set, batch_size=64, shuffle=False)

print(f"Train : {len(train_small)} (sous-échantillon de {len(train_full)})")
print(f"Test  : {len(test_set)}")
```

```python
#| eval: false
modele = modele.to(device)
# Seuls les paramètres de fc sont entraînés
optimizer = optim.Adam(modele.fc.parameters(), lr=1e-3)
loss_fn = nn.CrossEntropyLoss()

# Boucle d'entraînement rapide
import time
t0 = time.time()
for epoch in range(3):
    modele.train()
    # IMPORTANT : même en train(), les couches gelées restent gelées
    # Mais on doit garder les BN en eval pour éviter les fluctuations
    for module in modele.modules():
        if isinstance(module, nn.BatchNorm2d):
            module.eval()
    
    losses = []
    for Xb, yb in train_loader:
        Xb, yb = Xb.to(device), yb.to(device)
        logits = modele(Xb)
        loss = loss_fn(logits, yb)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
        losses.append(loss.item())
    
    # Eval
    modele.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for Xb, yb in test_loader:
            Xb, yb = Xb.to(device), yb.to(device)
            correct += (modele(Xb).argmax(dim=1) == yb).sum().item()
            total += yb.size(0)
    acc = correct / total
    print(f"Époque {epoch+1} | Loss {np.mean(losses):.4f} | Test acc {acc:.4f}")

print(f"\nTemps total : {time.time()-t0:.0f}s")
```

```python
#| eval: false
def entrainer_transfer(n_images, n_epochs=3):
    """Reproduit l'entraînement avec n_images."""
    # Modèle
    m = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
    for p in m.parameters():
        p.requires_grad = False
    m.fc = nn.Linear(m.fc.in_features, 10)
    m = m.to(device)
    
    # Sous-échantillon
    indices = np.random.choice(len(train_full), n_images, replace=False).tolist()
    loader = DataLoader(Subset(train_full, indices), batch_size=32, shuffle=True)
    
    # Training
    opt = optim.Adam(m.fc.parameters(), lr=1e-3)
    lf = nn.CrossEntropyLoss()
    
    for epoch in range(n_epochs):
        m.train()
        for module in m.modules():
            if isinstance(module, nn.BatchNorm2d):
                module.eval()
        for Xb, yb in loader:
            Xb, yb = Xb.to(device), yb.to(device)
            loss = lf(m(Xb), yb)
            opt.zero_grad(); loss.backward(); opt.step()
    
    # Eval
    m.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for Xb, yb in test_loader:
            Xb, yb = Xb.to(device), yb.to(device)
            correct += (m(Xb).argmax(dim=1) == yb).sum().item()
            total += yb.size(0)
    return correct / total

# Tester
tailles = [100, 500, 2000, 10000]
accs = []
for n in tailles:
    acc = entrainer_transfer(n)
    accs.append(acc)
    print(f"n={n:5d} images : accuracy test = {acc:.4f}")

# Graphique
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(tailles, accs, "o-", linewidth=2, markersize=10)
ax.set_xlabel("Nombre d'images train"); ax.set_ylabel("Accuracy test")
ax.set_xscale("log")
ax.axhline(0.8, color="red", linestyle="--", alpha=0.5, label="80% (utilisable)")
ax.set_title("Transfer learning : plus d'images = mieux, mais ça sature")
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#| eval: false
import time

def train_feature_extraction(n_epochs=5):
    m = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
    for p in m.parameters():
        p.requires_grad = False
    m.fc = nn.Linear(m.fc.in_features, 10)
    m = m.to(device)
    opt = optim.Adam(m.fc.parameters(), lr=1e-3)
    lf = nn.CrossEntropyLoss()
    
    t0 = time.time()
    for epoch in range(n_epochs):
        m.train()
        for module in m.modules():
            if isinstance(module, nn.BatchNorm2d):
                module.eval()
        for Xb, yb in train_loader:
            Xb, yb = Xb.to(device), yb.to(device)
            loss = lf(m(Xb), yb)
            opt.zero_grad(); loss.backward(); opt.step()
    temps = time.time() - t0
    
    m.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for Xb, yb in test_loader:
            Xb, yb = Xb.to(device), yb.to(device)
            correct += (m(Xb).argmax(dim=1) == yb).sum().item()
            total += yb.size(0)
    return correct / total, temps

def train_fine_tune(n_epochs=5):
    m = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
    for p in m.parameters():
        p.requires_grad = False
    for p in m.layer4.parameters():
        p.requires_grad = True
    m.fc = nn.Linear(m.fc.in_features, 10)
    m = m.to(device)
    opt = optim.Adam([
        {"params": m.layer4.parameters(), "lr": 1e-4},
        {"params": m.fc.parameters(), "lr": 1e-3},
    ])
    lf = nn.CrossEntropyLoss()
    
    t0 = time.time()
    for epoch in range(n_epochs):
        m.train()
        for Xb, yb in train_loader:
            Xb, yb = Xb.to(device), yb.to(device)
            loss = lf(m(Xb), yb)
            opt.zero_grad(); loss.backward(); opt.step()
    temps = time.time() - t0
    
    m.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for Xb, yb in test_loader:
            Xb, yb = Xb.to(device), yb.to(device)
            correct += (m(Xb).argmax(dim=1) == yb).sum().item()
            total += yb.size(0)
    return correct / total, temps

# Tester
print("Feature extraction :")
acc_fe, t_fe = train_feature_extraction()
print(f"  Acc {acc_fe:.4f} | Temps {t_fe:.0f}s")

print("\nFine-tuning :")
acc_ft, t_ft = train_fine_tune()
print(f"  Acc {acc_ft:.4f} | Temps {t_ft:.0f}s")
```

```python
#| eval: false
# Classes CIFAR-10 : 0=airplane, 1=auto, 2=bird, 3=cat, 4=deer, 5=dog, 6=frog, 7=horse, 8=ship, 9=truck
CLASSES_CIBLE = [2, 3, 5]  # bird, cat, dog
NOMS = ["bird", "cat", "dog"]

# Charger CIFAR-10 avec augmentation forte
transform_train_strong = transforms.Compose([
    transforms.Resize(224),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

train_full = torchvision.datasets.CIFAR10(
    root=os.path.join(DATA_ROOT, "cifar"), train=True, download=True, transform=transform_train_strong
)
test_full = torchvision.datasets.CIFAR10(
    root=os.path.join(DATA_ROOT, "cifar"), train=False, download=True, transform=transform_test
)

# Filtrer aux 3 classes
def filtrer_et_remaper(dataset, classes_cibles, n_par_classe=None):
    indices_par_classe = {c: [] for c in classes_cibles}
    for i, (_, y) in enumerate(dataset):
        if y in classes_cibles:
            indices_par_classe[y].append(i)
    
    # Sous-échantillonnage si demandé
    if n_par_classe:
        for c in classes_cibles:
            indices_par_classe[c] = indices_par_classe[c][:n_par_classe]
    
    tous = [i for inds in indices_par_classe.values() for i in inds]
    return tous

# Mapper classes 2,3,5 → 0,1,2
class RemappedDataset:
    def __init__(self, subset, classes_cibles):
        self.subset = subset
        self.mapping = {c: i for i, c in enumerate(classes_cibles)}
    def __len__(self):
        return len(self.subset)
    def __getitem__(self, idx):
        X, y = self.subset[idx]
        return X, self.mapping[y]

# Note : Cette implémentation peut nécessiter un ajustement selon version PyTorch
# L'important est de comprendre la démarche

# Train : 150 images (50/classe)
train_indices = filtrer_et_remaper(train_full, CLASSES_CIBLE, n_par_classe=50)
train_subset = Subset(train_full, train_indices)

# Test : toutes les images des 3 classes
test_indices = filtrer_et_remaper(test_full, CLASSES_CIBLE)
test_subset = Subset(test_full, test_indices)

print(f"Train : {len(train_subset)} images")
print(f"Test  : {len(test_subset)} images")

# DataLoaders (attention : classes originales, il faut remapper dans la loss)
train_loader = DataLoader(train_subset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_subset, batch_size=32)

# Modèle : ResNet-18 avec tête à 3 classes
modele = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
for p in modele.parameters():
    p.requires_grad = False
modele.fc = nn.Linear(modele.fc.in_features, 3)  # 3 classes
modele = modele.to(device)

# Training avec remap des labels en place
optimizer = optim.Adam(modele.fc.parameters(), lr=1e-3)
loss_fn = nn.CrossEntropyLoss()

label_to_idx = {c: i for i, c in enumerate(CLASSES_CIBLE)}

best_acc = 0
for epoch in range(10):
    modele.train()
    for m in modele.modules():
        if isinstance(m, nn.BatchNorm2d):
            m.eval()
    
    for Xb, yb in train_loader:
        # Remapper yb (2,3,5) → (0,1,2)
        yb_remap = torch.tensor([label_to_idx[int(y)] for y in yb], device=device)
        Xb = Xb.to(device)
        logits = modele(Xb)
        loss = loss_fn(logits, yb_remap)
        optimizer.zero_grad(); loss.backward(); optimizer.step()
    
    # Eval
    modele.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for Xb, yb in test_loader:
            yb_remap = torch.tensor([label_to_idx[int(y)] for y in yb], device=device)
            Xb = Xb.to(device)
            correct += (modele(Xb).argmax(dim=1) == yb_remap).sum().item()
            total += yb.size(0)
    acc = correct / total
    best_acc = max(best_acc, acc)
    print(f"Époque {epoch+1}: accuracy {acc:.4f}")

print(f"\n🎯 Meilleure accuracy : {best_acc:.4f}")
```
