```python
# Stats ImageNet : OBLIGATOIRE pour transfer learning
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD = [0.229, 0.224, 0.225]

transform_train = transforms.Compose([
    transforms.Resize(224),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.15, contrast=0.15),
    transforms.ToTensor(),
    transforms.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD)
])

transform_test = transforms.Compose([
    transforms.Resize(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD)
])

print("✅ Transforms définis")
```

```python
train_ds = ImageFolder(DATA_DIR / "train", transform=transform_train)
test_ds = ImageFolder(DATA_DIR / "test", transform=transform_test)

print(f"Train : {len(train_ds)} images")
print(f"Test  : {len(test_ds)} images")
print(f"Classes : {train_ds.classes}")
print(f"Mapping : {train_ds.class_to_idx}")

# DataLoaders
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True, num_workers=0)
test_loader = DataLoader(test_ds, batch_size=32, shuffle=False, num_workers=0)

# Shape d'un batch
for Xb, yb in train_loader:
    print(f"\nBatch : X {Xb.shape} | y {yb.shape}")
    print(f"Valeurs y : {yb.tolist()[:10]}...")
    break
```

```python
class SimpleCNN(nn.Module):
    def __init__(self, n_classes=3):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 32, 3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(64, 128, 3, padding=1)
        self.bn3 = nn.BatchNorm2d(128)
        self.pool = nn.MaxPool2d(2, 2)
        
        # Après 3 pools : 224 → 112 → 56 → 28
        self.fc1 = nn.Linear(128 * 28 * 28, 256)
        self.drop = nn.Dropout(0.3)
        self.fc2 = nn.Linear(256, n_classes)
    
    def forward(self, x):
        x = self.pool(F.relu(self.bn1(self.conv1(x))))
        x = self.pool(F.relu(self.bn2(self.conv2(x))))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = self.drop(x)
        return self.fc2(x)

def train_evaluer(modele, n_epochs=10, lr=1e-3):
    """Entraîne un modèle et renvoie accs + loss."""
    modele = modele.to(device)
    optimizer = optim.Adam(modele.parameters(), lr=lr)
    loss_fn = nn.CrossEntropyLoss()
    
    losses, accs = [], []
    for epoch in range(n_epochs):
        modele.train()
        epoch_losses = []
        for Xb, yb in train_loader:
            Xb, yb = Xb.to(device), yb.to(device)
            loss = loss_fn(modele(Xb), yb)
            optimizer.zero_grad(); loss.backward(); optimizer.step()
            epoch_losses.append(loss.item())
        losses.append(np.mean(epoch_losses))
        
        modele.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for Xb, yb in test_loader:
                Xb, yb = Xb.to(device), yb.to(device)
                correct += (modele(Xb).argmax(dim=1) == yb).sum().item()
                total += yb.size(0)
        accs.append(correct / total)
        print(f"Époque {epoch+1:2d}: loss {losses[-1]:.3f} | test acc {accs[-1]:.3f}")
    
    return modele, losses, accs

print("Training CNN from scratch...")
torch.manual_seed(42)
cnn_scratch = SimpleCNN()
cnn_scratch, losses_scratch, accs_scratch = train_evaluer(cnn_scratch, n_epochs=10)
print(f"\n🎯 Meilleure accuracy CNN from scratch : {max(accs_scratch):.3f}")
```

```python
def evaluer_details(modele, data_loader, classes):
    """Renvoie les prédictions + vrais labels."""
    modele.eval()
    tous_preds, tous_vrais = [], []
    with torch.no_grad():
        for Xb, yb in data_loader:
            Xb, yb = Xb.to(device), yb.to(device)
            preds = modele(Xb).argmax(dim=1)
            tous_preds.extend(preds.cpu().tolist())
            tous_vrais.extend(yb.cpu().tolist())
    return np.array(tous_vrais), np.array(tous_preds)

classes = train_ds.classes
y_true_scratch, y_pred_scratch = evaluer_details(cnn_scratch, test_loader, classes)

# Matrice de confusion
cm = confusion_matrix(y_true_scratch, y_pred_scratch)

fig, ax = plt.subplots(figsize=(7, 5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=classes, yticklabels=classes, ax=ax)
ax.set_xlabel("Prédit"); ax.set_ylabel("Vrai")
ax.set_title("Matrice de confusion — CNN from scratch")
plt.tight_layout()
plt.show()

# Rapport détaillé
print(classification_report(y_true_scratch, y_pred_scratch,
                             target_names=classes, digits=3))
```

```python
# Calculer les poids : inverse de la fréquence, normalisés
n_par_classe = np.array([32, 320, 64])  # fissure, ok, rayure (ordre ImageFolder)
freq = n_par_classe / n_par_classe.sum()
poids = 1.0 / freq
poids = poids / poids.sum() * len(poids)  # normaliser pour moyenne = 1
poids_t = torch.FloatTensor(poids).to(device)
print(f"Poids par classe : {dict(zip(classes, poids.round(3)))}")

# Entraînement avec loss pondérée
torch.manual_seed(42)
modele_w = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
for p in modele_w.parameters():
    p.requires_grad = False
modele_w.fc = nn.Linear(modele_w.fc.in_features, 3)
modele_w = modele_w.to(device)

optimizer = optim.Adam(modele_w.fc.parameters(), lr=1e-3)
loss_fn_w = nn.CrossEntropyLoss(weight=poids_t)

for epoch in range(10):
    modele_w.train()
    for module in modele_w.modules():
        if isinstance(module, nn.BatchNorm2d):
            module.eval()
    for Xb, yb in train_loader:
        Xb, yb = Xb.to(device), yb.to(device)
        loss = loss_fn_w(modele_w(Xb), yb)
        optimizer.zero_grad(); loss.backward(); optimizer.step()

# Évaluation détaillée
y_true_w, y_pred_w = evaluer_details(modele_w, test_loader, classes)

# Comparaison des rapports
print("=== Sans pondération (feature extraction) ===")
print(classification_report(y_true_tl, y_pred_tl, target_names=classes, digits=3))

print("\n=== Avec pondération ===")
print(classification_report(y_true_w, y_pred_w, target_names=classes, digits=3))
```

```python
# Analyse des erreurs : les plus "sûres"
modele_final.eval()
erreurs = []
with torch.no_grad():
    for Xb, yb in test_loader:
        Xb_gpu = Xb.to(device)
        probas = torch.softmax(modele_final(Xb_gpu), dim=1)
        preds = probas.argmax(dim=1)
        max_proba = probas.max(dim=1).values
        
        for i in range(Xb.size(0)):
            if preds[i].item() != yb[i].item():
                erreurs.append({
                    "img": Xb[i],
                    "vrai": classes[yb[i].item()],
                    "pred": classes[preds[i].item()],
                    "proba": max_proba[i].item()
                })

# Trier par confiance décroissante (erreurs les plus "confiantes")
erreurs.sort(key=lambda e: -e["proba"])

if erreurs:
    print(f"\nTotal erreurs : {len(erreurs)}")
    n_show = min(6, len(erreurs))
    fig, axes = plt.subplots(1, n_show, figsize=(3*n_show, 4))
    if n_show == 1:
        axes = [axes]
    for ax, err in zip(axes, erreurs[:n_show]):
        # Dénormaliser pour affichage
        img = err["img"].clone()
        for c in range(3):
            img[c] = img[c] * IMAGENET_STD[c] + IMAGENET_MEAN[c]
        img = img.clamp(0, 1).permute(1, 2, 0).numpy()
        ax.imshow(img)
        ax.set_title(f"Vrai: {err['vrai']}\nPréd: {err['pred']} ({err['proba']:.2%})",
                      fontsize=10, color="red" if err["pred"] != err["vrai"] else "green")
        ax.axis("off")
    plt.suptitle("Erreurs les plus confiantes (pistes d'amélioration)", y=1.02)
    plt.tight_layout()
    plt.show()
else:
    print("Aucune erreur ! 🎉")
```
