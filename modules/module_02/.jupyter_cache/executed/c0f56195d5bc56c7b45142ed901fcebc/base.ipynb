```python
np.random.seed(0)
data = np.random.randn(1000) * 10 + 50

# 1. Compter
n_sup_60 = np.sum(data > 60)  # True=1, False=0 → sum = nombre de True
print(f"Valeurs > 60 : {n_sup_60}")

# 2. Extraire
entre_40_60 = data[(data >= 40) & (data <= 60)]
print(f"Valeurs entre 40 et 60 : {len(entre_40_60)}")

# 3. Remplacer négatifs (il n'y en a probablement pas ici, mais le principe est là)
data[data < 0] = 0
print(f"Après remplacement, min = {data.min():.2f}")

# 4. Max et indice
max_val = data.max()
idx_max = np.argmax(data)
print(f"Max = {max_val:.2f} à l'indice {idx_max}")

# 5. Top 10 valeurs
# Deux méthodes
top10_sort = np.sort(data)[-10:]      # trier + prendre les 10 derniers
print(f"Top 10 (sort)      : {top10_sort}")

# Alternative plus efficace pour de gros arrays
top10_partition = np.partition(data, -10)[-10:]
top10_partition.sort()
print(f"Top 10 (partition) : {top10_partition}")
```

```python
np.random.seed(42)
notes = np.random.uniform(5, 20, size=(30, 5))
matieres = ['Maths', 'Français', 'Histoire', 'Anglais', 'Sport']

# 1. Moyenne par étudiant (on écrase les colonnes → axis=1)
moy_etudiants = notes.mean(axis=1)
print(f"Moyennes des 5 premiers étudiants : {moy_etudiants[:5].round(2)}")

# 2. Moyenne par matière (on écrase les lignes → axis=0)
moy_matieres = notes.mean(axis=0)
for nom, moy in zip(matieres, moy_matieres):
    print(f"  {nom:10s} : {moy:.2f}")

# 3. Matière avec meilleure moyenne
idx_best = np.argmax(moy_matieres)
print(f"\nMeilleure matière : {matieres[idx_best]} ({moy_matieres[idx_best]:.2f})")

# 4. Étudiants ≥ 12
n_bons = np.sum(moy_etudiants >= 12)
print(f"Étudiants à ≥12 : {n_bons}")

# 5. Meilleur étudiant
idx_meilleur = np.argmax(moy_etudiants)
print(f"Meilleur étudiant : n°{idx_meilleur} avec {moy_etudiants[idx_meilleur]:.2f}")

# 6. Matière la plus dispersée
stds = notes.std(axis=0)
idx_disperse = np.argmax(stds)
print(f"Matière la plus dispersée : {matieres[idx_disperse]} (std={stds[idx_disperse]:.2f})")
```

```python
np.random.seed(42)
n = 50
surfaces = np.random.uniform(20, 150, n)
occupants = np.random.randint(1, 6, n)
ages_batiment = np.random.randint(0, 80, n)
prix = (2500 * surfaces + 5000 * occupants - 500 * ages_batiment 
        + np.random.normal(0, 10000, n) + 50000)
prix[3] = 9_999_999
prix[17] = np.nan
surfaces[25] = 500

# 1. Masque NaN
has_nan = np.isnan(prix)
print(f"NaN dans prix : {has_nan.sum()}")

# 2. Imputation par la médiane
mediane_prix = np.nanmedian(prix)  # ignore les NaN
prix = np.where(has_nan, mediane_prix, prix)
print(f"Après imputation : {np.isnan(prix).sum()} NaN restants")

# 3. Détection outliers par IQR
def detecter_outliers_iqr(x):
    q1 = np.percentile(x, 25)
    q3 = np.percentile(x, 75)
    iqr = q3 - q1
    return (x < q1 - 1.5*iqr) | (x > q3 + 1.5*iqr)

out_prix = detecter_outliers_iqr(prix)
out_surf = detecter_outliers_iqr(surfaces)
print(f"Outliers prix     : {out_prix.sum()}")
print(f"Outliers surface  : {out_surf.sum()}")

# 4. Masque global (à GARDER = non outlier)
a_garder = ~(out_prix | out_surf)
print(f"Lignes à garder   : {a_garder.sum()} sur {n}")

# Appliquer le masque à toutes les variables
surfaces_clean = surfaces[a_garder]
occupants_clean = occupants[a_garder]
ages_clean = ages_batiment[a_garder]
prix_clean = prix[a_garder]

# 5. Construire X (on empile les colonnes)
X = np.column_stack([surfaces_clean, occupants_clean, ages_clean])
print(f"\nShape de X : {X.shape}")

# 6. Normalisation en UNE ligne grâce au broadcasting
X_norm = (X - X.mean(axis=0)) / X.std(axis=0)
print(f"Moyennes après normalisation : {X_norm.mean(axis=0).round(3)}")
print(f"Std après normalisation      : {X_norm.std(axis=0).round(3)}")

# 7. Corrélation surface/prix
corr = np.corrcoef(surfaces_clean, prix_clean)[0, 1]
print(f"\nCorrélation surface ↔ prix : {corr:.3f}")

# 8. Résumé final
print(f"\n=== Résumé ===")
print(f"Lignes finales : {X.shape[0]}")
print(f"Shape X_norm   : {X_norm.shape}")
```
